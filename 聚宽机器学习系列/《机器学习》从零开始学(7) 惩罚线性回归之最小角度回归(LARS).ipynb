本节目的： 进一步学习惩罚线性回归模型（前面已经学过其特例：岭回归）, 主要是理解惩罚线性回归模型的原理。

## 1) 惩罚线性回归的一般形式

首先写出惩罚回归的一般形式：

$$\boldsymbol{\beta}^* = \underset{\boldsymbol{\beta}}{\rm{argmin}} \left\{ \dfrac{1}{2 m}\sum_{i=1}^{m} {\left[y_i - \rm{pred}(\boldsymbol{x}_i, \boldsymbol{\beta})\right]^2} + \lambda \ \rm{pena}(\boldsymbol{\beta})\right\}$$

特别地，可写出惩罚线性回归的一般形式：

$$\beta_0^*, \boldsymbol{\beta}^* = \underset{\beta_0, \ \ \boldsymbol{\beta}}{\rm{argmin}} \left\{ \dfrac{1}{2 m}\sum_{i=1}^{m} {\left[y_i -  (\beta_0+ {\boldsymbol{x}_i}^T \boldsymbol{\beta})\right]^2} + \lambda \ \rm{pena}(\boldsymbol{\beta})\right\}$$

可以等价地改写成全向量表示：

$$\boldsymbol{\beta}^* = \underset{\boldsymbol{\beta}}{\rm{argmin}} \left\{\dfrac{1}{2 m} {\|\boldsymbol{y}-\boldsymbol{X} \ \boldsymbol{\beta}\|}^2+ \lambda \ \rm{pena}(\boldsymbol{\beta})\right\}$$

其中：$n$是属性个数， 而$m$是样本数， $\boldsymbol{\beta} = (\beta_0,\beta_1,\dots,\beta_n)^T,\quad \boldsymbol{X} = [x_{i \ j}]_{m \times (n+1)},  \quad x_{i \ 0}=1, \quad i=1,\dots,m, \quad j=0,1,\dots,n$


## 2) 几种特殊的惩罚线性回归

再特别一点，就是**岭回归**了：

$$\boldsymbol{\beta}^* = \underset{\boldsymbol{\beta}}{\rm{argmin}} \left\{ \dfrac{1}{2 m} {\|\boldsymbol{y}-\boldsymbol{X} \ \boldsymbol{\beta}\|}^2 + \dfrac{\lambda}{2} {\|\boldsymbol{\beta}\|}^2 \right\}$$

对应的优化目标函数$J(\boldsymbol{\beta})$的一阶导数方程：

$$0 = \nabla J(\boldsymbol{\beta}) = - \dfrac{1}{m} \boldsymbol{X}^T  (\boldsymbol{y}-\boldsymbol{X}  \boldsymbol{\beta}) +  \lambda \boldsymbol{\beta} $$

立刻可得到**岭回归**的封闭解：

$$\boldsymbol{\beta}^* = (\boldsymbol{X}^T \boldsymbol{X} + m \lambda \ \boldsymbol{I})^{-1} \boldsymbol{X}^T \boldsymbol{y}$$

直接将$\nabla J(\boldsymbol{\beta})$用在**梯度下降法**中也能很方便求出其数值解。

所以，岭回归的算法是很简单的。同时这个解的被求逆的部分在$\boldsymbol{X}^T \boldsymbol{X}$的主对角线元素增加$\lambda$，这本身就是一个著名的增强矩阵求逆数值稳定性的技巧。

同时可看出： **岭回归**就是配置了欧式距离平方惩罚项的线性回归。

类似地，**套索回归**(Lasso)就是配置了曼哈顿距离惩罚项的线性回归：

$$\rm{pena}(\boldsymbol{\beta}) = {\| \boldsymbol{\beta} \|}_1 = \sum_{j=1}^n \left| \beta_j \right|$$

遗憾的是，**套索回归** 没有封闭解，需要用数值方法来求解，并且尤其惩罚项的非线性，不好计算梯度（一阶导数），需要有点技巧。

当$\lambda$足够大的时候，**套索回归**的回归系数向量是稀疏的，而**岭回归**相比之下则是密集的。其原因可以从绘出惩罚值等高线直观感受到。关键是**套索回归**惩罚值等高线是折线，其转折角出现在坐标线上。而欧式距离平方惩罚项的等高线是曲线。

当$\lambda$足够小的时候，**套索回归**惩罚值等高线的“直线边”上可能成为最优点，进而其回归系数向量也是密集的，进而可以对回归系数进行排序，从而可看出属性的重要程度。

将岭惩罚项和套索惩罚项按比例混合，就是**ElasticNet惩罚项**:

$$\rm{pena}(\boldsymbol{\beta}) = (1-\alpha) \dfrac{1}{2} {\| \boldsymbol{\beta} \|}^2 + \alpha \ {\| \boldsymbol{\beta} \|}_1 \quad 0 < \alpha < 1$$

一般来说，是先确定$\alpha$值再确定$\lambda$值。

## 3) 求解惩罚线性回归

这里只介绍两种算法: **最小角度回归**（LARS）以及**Glmnet**。

虽然python中已经有成型的算法可以调用, 但是为了理解算法，这里依然展示了具体实现的代码。

### 理解最小角度回归于前向逐步回归的关系

LARS可以理解成一种改进的前向逐步回归算法。 从下面代码就能完全理解LARS算法的确切流程。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 获取红酒数据集
target_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(target_url, header=0, sep=";")
df.columns = ['非挥发性酸','挥发性酸','柠檬酸', '残留糖分', '氯化物', '游离二氧化硫','总二氧化硫', '密度', 
              'PH值', '亚硝酸盐', '酒精含量', '品质']

# 标准化,然后拆分成属性x和标签y
norm_df = (df - df.mean())/df.std()
xData = norm_df.values[:,:-1]; yData = norm_df.values[:,-1] 

# 步数 和 步长
nSteps = 350; stepSize = 0.004

# 回归系数\beta向量初始化
beta = np.array([0.0]*xData.shape[1])
# 存储每一步生成的回归系数\beta向量, 第0步就是全0向量
betaList = []
betaList.append(list(beta))

for i in range(nSteps):
    # 残差
    resid = yData - xData.dot(beta)

    # 计算属性与残差的相关性
    # 注意: 已标准化，所以此结果和真实相关性仅仅相差一个常数
    corr = resid.dot(xData)

    # 找出最大相关性(绝对值)属性
    iStar = np.abs(corr).argmax()
    corrStar = corr[iStar]

    # 更新beta (相关性为正值则增加一点；负值则减少一点)
    beta[iStar] += stepSize * corrStar / abs(corrStar)

    betaList.append(list(beta))

# 绘制每个属性的"步数:回归系数"曲线
plt.plot(betaList)
plt.xlabel("迭代步数")
plt.ylabel("系数值")
plt.show()
```

正如代码所示， 在惩罚线性回归中，标准化一般是第一步，确保不同属性具有相同的尺度。 算法包一般也会提供非标准化的选项，但很少听说不进行标准化有什么优势。

另外，我们注意到，我们没有计算$\beta_0$， 这是因为算法是基于标准化的变量上， 因此不需要“截距”$\beta_0$。

一般现代的机器学习算法，尤其是惩罚线性回归方法会生成多组解，这是因为模型中有待确认的参数，比如惩罚系数$\lambda$。  LARS虽然没有显式处理$\lambda$，但会有同样效果。 因为每步对应的$\beta$可以认为对应一个$\lambda$。 这说明尽管LARS算法没有明显出现惩罚项，但自带一个隐含的惩罚项，即对$\beta$向量空间具有某种意义下的约束。 所以，LARS算法无需明确指定惩罚项。

实际上，步数越小，模型越简单，对于的$\lambda$值越大，而上面的绘图结果表明对应的回归系数向量越稀疏。从这点来看，LARS很接近**套索回归**。

在前面的迭代中，回归系数离开0的过程可以看作属性重要的表现，更早离开0的属性是更重要的属性，据此可以生成属性重要性的排序。

**重点**： 生成属性重要性排序是惩罚线形回归模型的一个重要特征。使得该模型成为模型初步研究中的一个便捷工具，因为她可以帮助选择我们需要的属性值，这个过程也称作**特征工程**。


### 从数百个LARS生成的结果中选择最佳模型

代码中的注释已经能说明选择流程

```python

# 交叉验证的折数
nxval = 10

# 步数 和 步长
nSteps = 350; stepSize = 0.004

# 用了记录每一步在测试集上的残差，
# 注意：一共有（nxval*测试集样本数）个元素的数组。
errors = []
for i in range(nSteps):
    errors.append([])

for ixval in range(nxval):
    idxs = range(len(yData))
    idxTest = [i for i in idxs if i%nxval == ixval%nxval]
    idxTrain = [i for i in idxs if i%nxval != ixval%nxval]    
    xTest = xData[idxTest,:]; yTest = yData[idxTest]
    xTrain = xData[idxTrain,:]; yTrain = yData[idxTrain]
    
    # 复制上一小节的代码（但只针对训练集）
    beta = np.array([0.0]*xTrain.shape[1])
    betaList = []
    betaList.append(list(beta))
    for i in range(nSteps):
        resid = yTrain - xTrain.dot(beta)
        corr = resid.dot(xTrain)
        iStar = np.abs(corr).argmax()
        corrStar = corr[iStar]
        beta[iStar] += stepSize * corrStar / abs(corrStar)
        betaList.append(list(beta))
        
        # 填充测试集偏差矩阵（步数 * (nxval*测试集样本数)）
        err = yTest - xTest.dot(beta)
        errors[i].extend(list(err))

# 计算验证曲线（每步的均方误差）
cvCurve = []
for err in errors:
    cvCurve.append(np.dot(err,err)/len(err))

# 最优模型对应的步数（最小误差）
minStep = np.argmin(cvCurve)
minMse = cvCurve[minStep]
print ("最优模型对应的步数：", minStep)
print ("最优模型对应的均方误差：", minMse)

# 绘制交叉验证的均方误差曲线
plt.plot(cvCurve)
plt.xlabel("迭代步数")
plt.ylabel("均方误差")
plt.show()
```
