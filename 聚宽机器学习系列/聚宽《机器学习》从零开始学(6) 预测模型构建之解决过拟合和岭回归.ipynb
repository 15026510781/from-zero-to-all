#### 部署模型的性能模拟

前面已经提及了评估模型性能的标准流程：将数据集分成2部分，一部分用来进行训练（比如2/3），另一部分用来测试（比如1/3），最后评估模型性能只能基于测试集。

另一种预留数据的方法被称作**n折交叉验证**,具体做法是：先将数据集$n$等份, 然后进行$n$次遍历。 对于第$i\in (1,\dots,5)$次遍历, 用第$i$块数据作为测试集，其它数据作为训练集。

**n折交叉验证**的优点是，生成的模型能产生更低的泛化误差，具备更好的预测性能。缺点是要更多的训练时间。

选择测试样本时要**注意**： 选出的测试样本应该能够代表整个数据集。 如果研究对象有特殊的统计特征，抽样过程可能要更加小心。

### 4） 模型与数据的均衡

为了明确起见，我们成将**过拟合**概念限定为:**过拟合**是指训练集和测试集上的误差存在显著差异。  比如上一节用**最小二乘法**(OLS)解决岩石&水雷分类问题就存在过拟合。

本节的主题就是以**OLS**为基础，如何解决**过拟合**的问题。

其实，普通的最小二乘作为一个原型能很好地展示机器学习算法的方方面面。她是一个有监督的学习算法，包括训练过程以及测试过程。在某些方面可能会过拟合，但与其他现代的函数逼近算法都存在一些共性，当然OLS有一个重要缺点：当过拟合发生时，没有办法阻止学习过程。

本节通过**向前逐步回归**和**岭回归**两种方法来调整**普通最小二乘法**的瓶颈。

#### 通过权衡问题复杂度、模型复杂度和数据集规模来选择模型

我们的目标还是线性模型：

$$\boldsymbol{y}_{n \times 1} \sim \boldsymbol{X}_{n \times m} \boldsymbol{\beta}_{m \times 1} + \beta_0 \boldsymbol{1} = [\boldsymbol{1},\boldsymbol{X}]_{n \times (m+1)} \boldsymbol{\beta}_{(m+1) \times 1} $$

特别地，如果$n=m+1$，并且$[\boldsymbol{1},\boldsymbol{X}]_{n \times n}$可逆的话，于是符号$\sim$可以替换为$=$,这意味着所得到的$\boldsymbol{\beta}_{n \times 1}$可以精确拟合标签, 于是出现了过拟合。

过拟合的根源在于有太多的列（属性）,进而又太多的参数，自然而然想到的一个蛮力方法就是去掉一些列，这种方法就是**最佳子集选择**。

#### 使用前向逐步回归来控制过拟合

最蛮力的**最佳子集选择**就是： 选择$i$个属性子集，找出就佳的模型。只要循环$i = 1,\dots$,就能找出**最佳子集**及其对应的模型。

当然这种蛮力需要大量计算，于是有了简化版的**前向逐步回归**：首先从1列子集出发，找到效果最好的那一列属性，然后寻找与其组合效果最佳的第2个属性（无须评估所有的2列子集）, ....

下面以红酒品质数据集来展示**前向逐步回归**方法：

首先，获取数据，并将数据集拆分成**测试数据集**和**训练数据集**

```python
import numpy as np
import pandas as pd

target_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(target_url, header=0, sep=";")
df.columns = ['非挥发性酸','挥发性酸','柠檬酸', '残留糖分', '氯化物', '游离二氧化硫','总二氧化硫', '密度', 
              'PH值', '亚硝酸盐', '酒精含量', '品质']

test_df = df.loc[df.index%3==0]
train_df = df.loc[df.index%3!=0]

xTrain = train_df.values[:,:-1]; yTrain = train_df.values[:,-1]
xTest = test_df.values[:,:-1]; yTest = test_df.values[:,-1]
```

然后，就是用前向逐步回归方法来得到最佳模型.

*** 前向逐步回归机制：找到最好的一列（记为1），然后把其他列往里加，然后找到与1搭配最好的一列组成【1，2】，以此类推

```python
import matplotlib.pyplot as plt
from sklearn import linear_model

def modelRMSE(idxs):
    '''
        对指定属性子集建立预测模型并计算测试集误差
    '''
    
    _xTrain = xTrain[:,idxs]; _xTest = xTest[:,idxs]
    model = linear_model.LinearRegression()
    model.fit(_xTrain,yTrain)
    return model,np.linalg.norm(yTest- model.predict(_xTest),2)/np.sqrt(len(yTest))

all_idxs = range(xTrain.shape[1])
idxset = []
errs = []
for i in all_idxs:
    for i in all_idxs:    
    #遍历评估后的误差列表
        errList = []
    
        # 遍历剩余属性集
        tmpSet = list(set(all_idxs) - set(idxset))
        for att in tmpSet:
            # 待评估的属性集
            idxs = [] + idxset
            idxs.append(att)
        
            _, _err = modelRMSE(idxs)
            errList.append(_err)
    
    # 寻找最好模型
    iBest = np.argmin(errList)
    idxset.append(tmpSet[iBest])
    errs.append(errList[iBest])

print ("最佳样本外误差：", errs)
print ("最佳属性索引：", idxset)    
print ("最佳属性名：", )
for idx in idxset:
    print (df.columns[idx])
print ("")

idx = np.argmin(errs)
print ("最佳属性集大小：%s"%(idx+1))
print ("最佳属性集大小：",)
for i in range(idx+1):
    print (df.columns[idxset[i]])
print ("")

# 属性-误差曲线
plt.plot(errs, 'k')
plt.xlabel("属性个数")
plt.ylabel("误差(RMS)")
plt.show()
```

由误差图可见，随着属性集个数增大，误差降低，直到9。

#### 评估并理解你的预测模型

绘制预测结果和实际结果的散点图

```python
_xTrain = xTrain[:,idxs]; _xTest = xTest[:,idxs]
model,_ = modelRMSE(idxset)
plt.scatter(model.predict(_xTest), yTest, s=100, alpha=0.1)
plt.xlabel("预测口感得分")
plt.ylabel("实际口感得分")
plt.show()
```

```python
plt.hist(yTest-model.predict(_xTest))
plt.xlabel("误差")
plt.ylabel("次数")
plt.show()
```

```python
# 选择alpha列表（一般按10的倍数递减）
alphaList = [0.1**i for i in range(7)]

rmsError = []
for alph in alphaList:
    model = linear_model.Ridge(alpha=alph)
    model.fit(xTrain, yTrain)
    rmsError.append(np.linalg.norm(yTest-model.predict(xTest), 2) / np.sqrt(len(yTest)))


print ("alpha对应的RMS误差：",)
for i in range(len(rmsError)):
    print ("(%s,%s)"%(alphaList[i],rmsError[i]))
print ("")

# alpha-误差曲线
plt.plot(rmsError, 'k')
plt.xlabel("-log(alpha)")
plt.ylabel("误差(RMS)")
plt.show()

# 最佳模型
idx = np.argmin(rmsError)
alphBest = alphaList[idx]
print ("最佳alpha：", alphBest)
model = linear_model.Ridge(alpha=alphBest)
model.fit(xTrain, yTrain)

# 预测 与 实际 的散点图
plt.scatter(model.predict(xTest), yTest, s=100, alpha=0.1)
plt.xlabel("预测口感得分")
plt.ylabel("实际口感得分")
plt.show()

# 误差直方图
plt.hist(yTest-model.predict(xTest))
plt.xlabel("误差")
plt.ylabel("次数")
plt.show()
```
