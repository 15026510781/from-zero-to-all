## 3) 逻辑回归与惩罚逻辑回归

另一种使用惩罚线性回归模型进行分类的方法是使用**惩罚逻辑回归**，当然在此之前需要理解**逻辑回归**。 当然，按照我的习惯套路，就需要理解其背后的数学原理。

### 逻辑回归(Logistic Regression)

**逻辑回归**是以概率模式识别为目标，用线性回归作为工具的模型。 所谓**概率模式识别**就是对与模式$\boldsymbol{x}$对应的类别$y$的**后验概率**$P(y \ | \ \boldsymbol{x})$进行学习，其所属类别为后验概率最大所对应的类别。

$$ \hat{y} = \underset{y = 1, \dots, c}{\rm{argmin}} \ P(y \ | \ \boldsymbol{x})$$

当然，概率是无法用线性模型近似，需要作一个**对数变换**$[0,1] \to \mathbb{R}$ (以**二分类问题**为例)：

$$\ln \left( \dfrac{P(y \ | \ \boldsymbol{x})}{1 - P(y \ | \ \boldsymbol{x})} \right)$$

进而可以和线性回归模型建立联系(引入了一个假设)：

$$\ln \left( \dfrac{P(y \ | \ \boldsymbol{x})}{1 - P(y \ | \ \boldsymbol{x})} \right) = \boldsymbol{x}^T \boldsymbol{\beta} \Longrightarrow P(y \ | \ \boldsymbol{x}) = \dfrac{1}{1 + \exp (-  \boldsymbol{x}^T \boldsymbol{\beta})}$$ 

其中：$\boldsymbol{x} = (1,x_1,\dots,x_n)^T, \quad \boldsymbol{\beta} = (\beta_0,\beta_1,\dots, \beta_n)$。

在给定数据集${\{(\boldsymbol{x}_i,y_i) \}}_{i=1}^m$中，线性回归模型中用**均方误差**来度量其性能，但在逻辑回归模型中需要用**对数拟然**(log-likelihood)最大化来度量其性能：

$$
\ln \prod_{i=1}^m {P(y_i \ | \ \boldsymbol{x}_i)^{y_i} (1 - P(y_i \ | \ \boldsymbol{x}_i))^{(1-y_i)}} = \boldsymbol{y}^T \boldsymbol{X} \boldsymbol{\beta} - \sum_{i=1}^m {\left[ \ln (1 + \exp(\boldsymbol{x}_i^T \boldsymbol{\beta})) \right]}
$$

其中：其中$\boldsymbol{X} = (\boldsymbol{x}_1,\dots \boldsymbol{x}_m)^T$，而$m$是数据集的样本数。 

按惯例，转换成最小化问题，就是**不含惩罚项的逻辑回归问题**：

$$
L(\boldsymbol{\beta}) =  \sum_{i=1}^m {\left[ \ln (1 + \exp(\boldsymbol{x}_i^T \boldsymbol{\beta})) \right]} - \boldsymbol{y}^T \boldsymbol{X} \boldsymbol{\beta}, \quad X = (\boldsymbol{x}_1,\dots, \boldsymbol{x}_m)^T
$$

$$
\beta^* = \underset{\boldsymbol{\beta}}{\rm{argmin}} \ L(\boldsymbol{\beta})
$$

### 迭代重加权最小二乘法(IRLS)

**IRLS方法**也叫Newton-Raphson，我们可以用这个方法求解**惩罚逻辑回归**问题：

$$
J(\boldsymbol{\beta}) = \dfrac{1}{m} L(\boldsymbol{\beta}) +  (1-\alpha) \dfrac{\lambda}{2}  {\| \boldsymbol{\beta} \|}^2 + \alpha \lambda  {\| \boldsymbol{\beta} \|}_1
$$

$$
\beta^* = \underset{\boldsymbol{\beta}}{\rm{argmin}} \ J(\boldsymbol{\beta})
$$


类似第8节的处理方式， 可以在当前迭代解$\tilde{\boldsymbol{\beta}}$附近对$L(\boldsymbol{\beta})$进行近似泰勒展开：

$$
L(\boldsymbol{\beta}) \simeq  \hat{L}(\boldsymbol{\beta}) = L(\tilde{\boldsymbol{\beta}}) + \boldsymbol{g}^T (\boldsymbol{\beta} - \tilde{\boldsymbol{\beta}}) + \dfrac{1}{2}  (\boldsymbol{\beta} - \tilde{\boldsymbol{\beta}})^T \boldsymbol{H} (\boldsymbol{\beta} - \tilde{\boldsymbol{\beta}})
$$

其中：

$$
\boldsymbol{g} = \nabla L(\tilde{\boldsymbol{\beta}}) = \boldsymbol{X}^T (\boldsymbol{\mu} - \boldsymbol{y} ), \quad  \boldsymbol{\mu} = \dfrac{1}{1 + \exp (-\boldsymbol{X}\tilde{\boldsymbol{\beta}})}
$$

$$
\boldsymbol{H} = {\nabla}^2 L(\tilde{\boldsymbol{\beta}}) = \boldsymbol{X}^T \boldsymbol{W} \boldsymbol{X}, \quad \boldsymbol{W} = \rm{diag}(\{\mu_i (1-\mu_i)\}) 
$$

泰勒展开近似后，容易算出$\hat{L}(\boldsymbol{\beta})$的局部最优封闭解（无惩罚项）：

$$
\begin{array}
&\hat{\boldsymbol{\beta}} &=& \underset{\boldsymbol{\beta}}{\rm{argmin}}  \ \hat{L}(\boldsymbol{\beta}) = \tilde{\boldsymbol{\beta}} - \boldsymbol{H}^{-1} \boldsymbol{g} \\ 
&=& \left(\boldsymbol{X}^T \boldsymbol{W} \boldsymbol{X} \right)^{-1} \boldsymbol{X}^T \boldsymbol{W} \boldsymbol{z}, \quad \boldsymbol{z} = \boldsymbol{X}\tilde{\boldsymbol{\beta}} + \boldsymbol{W}^{-1}(\boldsymbol{y} - \boldsymbol{\mu}) \\
&=& \tilde{\boldsymbol{\beta}} + \boldsymbol{H}^{-1} \boldsymbol{X}^T \boldsymbol{W} \boldsymbol{r}, \quad \boldsymbol{r} =\boldsymbol{z} -\boldsymbol{X} \tilde{\boldsymbol{\beta}}  
\end{array}
$$

----------------------------

**【题外话】**

我们注意到**加权最小二乘法**也有类似$\hat{\boldsymbol{\beta}}$的解：

$$
\beta^* = \underset{\boldsymbol{\beta}}{\rm{argmin}} \ \left\{(\boldsymbol{y}-\boldsymbol{X} \boldsymbol{\beta})^T\boldsymbol{W}(\boldsymbol{y}-\boldsymbol{X} \boldsymbol{\beta})  \right\} = \left(\boldsymbol{X}^T \boldsymbol{W} \boldsymbol{X} \right)^{-1} \boldsymbol{X}^T \boldsymbol{W} \boldsymbol{y}
$$

但是我们这里的$\boldsymbol{W}$在是迭代过程中变化的，所以称作**迭代重加权最小二乘法**(IRLS)。

----------------------------

有了这个局部最优解$\hat{\boldsymbol{\beta}}$，我们可以将$\hat{L}(\boldsymbol{\beta})$改写, 也就是把$J(\boldsymbol{\beta})$近似为：

$$
J(\boldsymbol{\beta}) \simeq \tilde{J}(\boldsymbol{\beta}) = \dfrac{1}{2 m} (\boldsymbol{\beta} - \hat{\boldsymbol{\beta}})^T \boldsymbol{H} (\boldsymbol{\beta} - \hat{\boldsymbol{\beta}}) +  (1-\alpha) \dfrac{\lambda}{2}  {\| \boldsymbol{\beta} \|}^2 + \alpha \lambda  {\| \boldsymbol{\beta} \|}_1 
$$

然后，将套用第8节的方法(**Glmnet算法**)，容易推导出$\tilde{J}(\boldsymbol{\beta})$的局部最优解（含惩罚项）：

$$
\tilde{\boldsymbol{\beta}} = \underset{\boldsymbol{\beta}}{\rm{argmin}}  \ \tilde{J}(\boldsymbol{\beta}) = \dfrac{S( \dfrac{\boldsymbol{H}}{m}\hat{\boldsymbol{\beta}}, \alpha \lambda)}{\dfrac{\boldsymbol{H}}{m}+(1-\alpha)\lambda}
$$

其中，$S(*,*)$就是第8节的**套索缩减函数**。 在具体算法中$\boldsymbol{H}$可简化为对角矩阵$\rm{diag}(H_{j,j})$。

如此反复迭代，直到满足精度。


**【IRLS单步迭代流程总结】**

首先是可以向量化计算的部分（**同步更新**）：

$$
\begin{array}
& \boldsymbol{\eta} = \boldsymbol{X} \boldsymbol{\beta} \\
\boldsymbol{\mu} = \dfrac{1}{1+\exp(\boldsymbol{-\eta})} \\
\boldsymbol{w} = \left\{ \mu_i(1-\mu_i) \right\}_{i=1,\dots,m} \\
\boldsymbol{W} = \rm{diag}(w) \\
\boldsymbol{z} = \boldsymbol{\eta} + \boldsymbol{W}^{-1}(\boldsymbol{y} - \boldsymbol{\mu})\\
\boldsymbol{H} = \boldsymbol{X}^T \boldsymbol{W} \boldsymbol{X} 
\end{array}
$$

然后就是逐个系数计算的部分（**异步更新**）：

$$
\begin{array}
& \boldsymbol{r} = \boldsymbol{z} - \boldsymbol{X} \boldsymbol{\beta} \\
\beta_j \leftarrow \beta_j + \dfrac{\left( \boldsymbol{X}^T \boldsymbol{W} \boldsymbol{r} \right)_j}{H_{j,j}} \\
\beta_j \leftarrow \dfrac{S(\dfrac{H_{j,j}}{m} \beta_j, \alpha \lambda)}{\dfrac{H_{j,j}}{m}+(1-\alpha)\lambda}
\end{array}, \quad j=1,\dots,n
$$

**【注意】** 

如果仅仅从数学推导看，可直接写成$\boldsymbol{r} = \boldsymbol{W}^{-1}(\boldsymbol{y} - \boldsymbol{\mu})$，但本节写成了两步$\boldsymbol{r} = \boldsymbol{z} - \boldsymbol{X} \boldsymbol{\beta}$和$\boldsymbol{z} = \boldsymbol{\eta} + \boldsymbol{W}^{-1}(\boldsymbol{y} - \boldsymbol{\mu})$。 形式上看虽然并没有本质区别，但算法上有本质区别的：$\boldsymbol{z}$的计算是采用**同步更新**,而$\boldsymbol{r}$的计算时采用**异步更新**。

---------------------------

下面是上述算法的忠实实现，比原书的代码简洁了很多，并且块了N多倍。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def S(z,gamma):
    if gamma >= np.abs(z):
        return 0.0
    if z > 0.0:
        return z - gamma
    else:
        return z + gamma

def glmnetIRLS(xx, y, beta, lam, alp):
    '''
        Glmnet和IRLS两个算法结合在一起，用于解决惩罚逻辑回归问题. 这是一个单步迭代公式
    参数：
        xx:  属性矩阵X
        y:   标签向量
        beta: 回归系数向量(更新后返回)
        lam:  \lambda
        alp:  \alpha
    '''

    m,n = xx.shape
       
    eta = np.dot(xx, beta) # = X beta
    eta[eta<-100] = -100
    
    mu = 1.0/(1.0+np.exp(-eta))
    
    w = mu * (1.0 - mu)
    
    cond1 = (np.abs(mu)<1e-5)
    cond2 = (np.abs(1.0-mu)<1e-5)
    mu[cond1] = 0.0
    mu[cond2] = 1.0
    w[cond1|cond2] = 1e-5
    
    ww = np.diag(w)
    
    z = (y-mu)/w + eta # = eta + W^{-1} (y - mu)
    
    for j in range(n):
        r = z - np.dot(xx,beta) # = z - X beta
        
        sumWxr = (w*xx[:,j]*r).sum()  # = (X^T W r)[j]
        sumWxx =  (w*xx[:,j]*xx[:,j]).sum() # = (X^T W X)[j,j]
        
        beta[j] = beta[j] + sumWxr / sumWxx
        
        if j > 0:
            avgWxx = sumWxx / m
            beta[j] = S(beta[j]*avgWxx, lam * alp) / (avgWxx + lam * (1.0 - alp))
    
    return beta

target_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data"
df = pd.read_csv(target_url, header=None, prefix='V')

# 分类标签数值化
df['V60'] = df.iloc[:,-1].apply(lambda v: 1.0 if v=='M' else 0.0)

# 切分属性和标签 然后 标准化 
xData = df.values[:,:-1]; yData = df.values[:,-1] 

# 只对X正规化, y只计算均值和标准差
xData = (xData - xData.mean(axis=0))/xData.std(axis=0)
yMean = yData.mean();  yStd = yData.std()

# 选择alpha参数
alpha = 0.8  # ElasticNet回归

# 确定lambda初始值：导致所有beta值都为零的lambda值（最简模型）
lam = (np.abs(yData.dot(xData))/xData.shape[0]).max()/alpha

# 扩展X成[1,X]
xData = np.concatenate((np.ones((xData.shape[0],1)), xData), axis=1)
nrow, ncol = xData.shape

# beta初始化
beta = np.zeros(ncol)
beta[0] = np.log(yMean/(1-yMean))

# 记录历史beta
betaMat = []
betaMat.append(list(beta[1:]))
beta0Mat = []
beta0Mat.append(beta[0])

# 迭代步数
nSteps = 100

# lambda缩减乘子：Fredman建议每步迭代后都要稍微减小lambda
lamMult = 0.93 # Fredman建议: lamMult^nSteps = 0.001

# 记录属性回归系数变成非零先后次序
nzList = []

# 开始进行lam迭代计算
for iStep in range(nSteps):
    lam = lam * lamMult
    
    deltaBeta = 100.0
    eps = 0.01
    iterStep = 0
    
    # 开始Glmnet算法迭代
    while deltaBeta > eps:
        iterStep += 1
        if iterStep > 100: #100:
            break
        
        # 上一步beta
        _beta = beta.copy()

        # IRL算法递归公式
        beta = glmnetIRLS(xData, yData, beta, lam, alpha)
        
        # 计算精度
        deltaBeta = np.abs(beta[1:]-_beta[1:]).sum() / np.abs(beta[1:]).sum()
    
    # 记录beta历史
    betaMat.append(list(beta[1:]))
    beta0Mat.append(beta[0])

    # 记录回归系数刚好变成非零的属性
    nzBeta = np.where(beta[1:]!=0)[0]
    for q in nzBeta:
        if (q in nzList) == False:
            nzList.append(q)

# 打印属性重要性排序
for idx in nzList:
    print (df.columns[idx],)

# lambda-误差曲线
plt.plot(betaMat)
plt.xlabel(u"迭代步数")
plt.ylabel(u"系数值")
plt.show()
```
