## 1）距离度量

我们所谓的**距离**是比**欧几里得距离**更一般、更抽象的概念，又叫**度量**。 在任意非空集合$\Omega$中，如果一个映射$d:\Omega \times \Omega \rightarrow \mathbb{R}$满足：

1) 半正定性：$d(x,y) \ge 0 \quad \mathrm{and} \quad  d(x,y)=0 \Leftrightarrow x=y, \quad \forall x,y\in\Omega$;

2) 对称性： $d(x,y) = d(y,x), \quad \forall x,y\in\Omega$;

3) 三角不等式： $d(x,z) \le d(x,y) + d(y,z), \quad \forall x,y,z \in\Omega$.

那么我们称$d$是$\Omega$的**距离**或**度量**, 而$(\Omega,d)$则称作**度量空间**。

下面是几个常用的距离（但不限于）：

- **闵可夫斯基距离**(具备平移不变性和尺度不变性)

$$
d(\boldsymbol{x},\boldsymbol{y}) = \left( \sum_{i=1}^n \left| x_i - y_i \right|^p \right)^{\frac{1}{p}}, \quad p \ge 1
$$

特别地， 当$p=2$时，就是**欧式距离**(闵可夫斯基距离中唯一具有旋转不变性的距离):

$$
d_2(\boldsymbol{x},\boldsymbol{y}) = \| \boldsymbol{x} - \boldsymbol{y} \|  = \sqrt{\sum_{i=1}^n \left( x_i - y_i \right)^2 } 
$$

当$p=1$时，就是**曼哈顿距离**:

$$
d_1(\boldsymbol{x},\boldsymbol{y}) = \| \boldsymbol{x} - \boldsymbol{y} \|_1  = \sum_{i=1}^n \left| x_i - y_i \right| 
$$

当$p \rightarrow \infty$时，就是**曼切比雪夫距离**:

$$
d_{\infty}(\boldsymbol{x},\boldsymbol{y}) = \| \boldsymbol{x} - \boldsymbol{y} \|_{\infty}  = \underset{i}{\mathrm{max}} \left| x_i - y_i \right| 
$$

- **海明距离**

$$
d_0(\boldsymbol{x},\boldsymbol{y}) = \| \boldsymbol{x} - \boldsymbol{y} \|_0  = \sum_{i=1}^n \left( x_i - y_i \right)^0
$$

- **椭圆距离**(又称**马氏距离**, 欧氏距离可以看成是**圆距离**是其特例)

$$
d_M(\boldsymbol{x},\boldsymbol{y}) = \sqrt{(\boldsymbol{x} - \boldsymbol{y})^T \boldsymbol{\Sigma}^{-1} (\boldsymbol{x} - \boldsymbol{y})} 
$$

其中，$\boldsymbol{\Sigma}$是数据集协方差矩阵，而$\boldsymbol{M} = \boldsymbol{\Sigma}^{-1}$对应一个超椭球面$\boldsymbol{x}^T \boldsymbol{M} \boldsymbol{x} = r $。


## 2）原型聚类

**聚类**是个操作，试图将数据集的样本划分成若干不相交的子集。 每个子集称为一个**簇**，这样的子集组成的集合称为**簇划分**。 所以，**聚类**就是试图对样本集进行**簇划分**。

**原型**是指样本空间中具有代表性的点, 又称为**范例**。 而**原型聚类**，就是首先假设聚类结构能够通过一组原型来刻画，然后将这组原型找出来。

下面是几组著名的原型聚类算法

### k均值算法

这是一个无监督学习算法。

这个算法的目标是：对给定样本集$\left\{ \boldsymbol{x}_i \right\}_1^m$进行**k阶最优簇划分**$C=\left\{ C_1,\dots,C_k \right\}$：

$$
C^* = \underset{C}{\mathrm{argmin}} \ E(C), \quad E(C)=\sum_{i=1}^k \sum_{\boldsymbol{x} \in C_i} \ \|\boldsymbol{x} - \boldsymbol{\mu}_i \|^2, \quad  \boldsymbol{\mu}_i = \dfrac{1}{|C_i|} \sum_{\boldsymbol{x} \in C_i} \ \boldsymbol{x}
$$

其中的均值点$\boldsymbol{\mu}_i$就是一个**原型**或**范例**。这个优化问题的目标函数$E(C)$，直观上看，刻画了簇内样本围绕簇均值的紧密程度，$E(C)$值越小簇内样本的相似度就越高。

可惜这是一个NP难问题， **k均值算法**则采用贪心策略近似求解这个优化问题：

![k均值算法](https://jiage.la/upload/201709213424325104810.png)

下面是用Python实现**k均值算法**(充分利用了向量化计算)。

```python
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt

# 这里选择sklearn提供样本生成器
centers = [[1, 1], [-1, -1], [1, -1]]
X, _ = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=0)
m,n = X.shape

# 聚类数
k = 3

# 【注意】： 本算法无需参考样本标签值 labels

# 随机选出k个样本作为初始均值点
idx = list(range(m))
np.random.seed(0); np.random.shuffle(idx)
Mu = X[idx[:k]]   # k * n矩阵

count = 0
while count < 1000:
    count += 1
    
    # 根据均值进行簇划分：将每个样本划分到距离最近的那个均值所在的簇(采用欧几里得距离)
    
    # 计算第i个样本到第j个均值点的距离D_{i,j}所构成的 m*k矩阵 (实际计算中用D代表D^2不影响结果)
    # D_{i,j} = (x_i-mu_j)^T (x_i-mu_j) = x_i^T x_i - 2 x_i^T mu_j + mu_j^T mu_j
    # i = 1,...,m
    # j = 1,...,k
    D = ((X**2).sum(axis=1) - 2*np.dot(Mu,X.T)).T + (Mu**2).sum(axis=1)
    
    # 求出每个样本所对应的最近均值点索引
    idxMinD = D.argmin(axis=1)
    
    # 簇划分(记录索引)
    C = [list(np.where(idxMinD==j)[0]) for j in range(k)]
    
    # 计算新的簇内均值点
    Mu_new = np.array([X[C[j]].mean(axis=0) for j in range(k)])
    
    # 如果簇内均值点无变化，则跳出循环
    if (Mu_new - Mu == 0).prod():
        break
     
    # 更新簇内均值点
    Mu = Mu_new 

print (u"迭代次数:",count)

# 根据簇划分C，对样本进行分类
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, k)]
for j in range(k):
    X_j = X[C[j]]
    plt.plot(X_j[:, 0], X_j[:, 1], 'o', markerfacecolor=colors[j], markeredgecolor='k', markersize=6)
# 绘制均值点
plt.plot(Mu[:,0], Mu[:,1], 'r*', markersize=18)
plt.xticks(())
plt.yticks(())
plt.show()
```

## 3）密度聚类

**密度聚类**，假设聚类结构能够通过样本分布的紧密程度确定，通常情况下，是从样本密度的角度考察样本之间的可连接性，并基于可连接样本不断扩展聚类簇已获得最终的聚类结果。

**DBSCAN**(Density-Based Spatial Clustering of Applications with Noise)是一种著名的密度聚类算法： 给定**邻域半经**$\epsilon$和**邻居样本数阈值**$\mathrm{minPts}$，如果一个样本$\boldsymbol{x}$的$\epsilon$-邻域中至少有$\mathrm{minPts}$个邻居样本，则称$\boldsymbol{x}$为**核心样本**。

如果样本$\boldsymbol{x}_j$位于某个核心样本$\boldsymbol{x}_i$的邻域中，则称$\boldsymbol{x}_j$由$\boldsymbol{x}_i$**密度直达**。

如果在样本$\boldsymbol{x}_j=\boldsymbol{p}_1$和核心样本$\boldsymbol{x}_i=\boldsymbol{p}_K$之外，可以找到一系列其它的核心样本$\left\{\boldsymbol{p}_k\right\}_2^{K-1}$，并且满足$\boldsymbol{p}_{k+1}$由$\boldsymbol{p}_{k}$**密度直达**，则称$\boldsymbol{x}_j$由$\boldsymbol{x}_i$**密度可达**。

对样本$\boldsymbol{x}_i$和样本$\boldsymbol{x}_j$，存在核心样本$\boldsymbol{x}_k$，$\boldsymbol{x}_i$和$\boldsymbol{x}_j$都由$\boldsymbol{x}_k$**密度可达**，则称$\boldsymbol{x}_i$与$\boldsymbol{x}_j$**密度相连**。

![DBSCAN基本概念](https://jiage.la/upload/201709213450045892328.png)

这个算法的核心： 由任意一个**核心样本**出发，找出所有**密度可达**的样本，然后聚类成簇。

![DBSCAN算法](https://jiage.la/upload/201709213454537340166.png)

下面就是这个算法的Python代码的实现。

```python
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt

# 这里选择sklearn提供样本生成器
centers = [[1, 1], [-1, -1], [1, -1]]
X, _ = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=0)
m,n = X.shape

# 邻域参数(最好先对数据进行分析，估选合适的参数)
eps = 0.4
minPts = 30

'''
    确定核心对象集
'''

# 计算第i个样本到第j个样本的距离D_{i,j}，进而构成m*m距离矩阵D
# D_{i,j} = (x_i-x_j)^T (x_i-x_j) = x_i^T x_i - 2 x_i^T x_j + x_j^T x_j
# i = 1,...,m
# j = 1,...,m
# D = sqrt(D)
XX = np.dot(X,X.T)
D = (XX.diagonal()-2*XX).T + XX.diagonal()
D = np.sqrt(D)

# 计算各样本的eps-邻域数量
pts = (D<eps).sum(axis=1)

# 最后选择“邻域数量 >= minPts + 1”的样本作为核心样本集（邻域数量包括了样本到自身的距离，所以要+1）
idxOmega = list(np.where(pts >= minPts+1)[0])

'''
    以任意核心对象为出发点，找出由其密度可达的样本生成一个聚类簇
    遍历完所有的核心对象， 就可以生成簇划分
'''

# 聚类簇数初始化
k = 0

# 初始化未访问样本集合(索引)
idxGamma = range(m) 
idxC = []

# 大循环： 首先随机选取核心样本, 最后还要剔除同类核心样本
np.random.seed(0)
while len(idxOmega)>0:
    # 记录当前未访问样本集(复制)
    idxOldGamma = list(idxGamma)
    # 首先随机选取核心样本
    idxQ = []
    o_id = np.random.choice(idxOmega,1)[0]
    
    # 加入初始化队列Q  并从Gamma中剔除
    idxQ.append(o_id)
    idxGamma = list(set(idxGamma) - set([o_id]))
    while len(idxQ)>0:
        # 从队列Q中选取首个样本
        q_id = idxQ.pop(0)
        
        if pts[q_id] >= minPts + 1:
            # q的邻域样本（不包含自身）
            ne = list(np.where((D<eps)[q_id])[0])
            ne.remove(q_id)
            
            # q的邻域样本与未访问样本Gamma的交集
            idxDelta = list(set(ne).intersection(idxGamma))
            
            # 将Delta加入队列Q
            idxQ.extend(idxDelta)
            
            # 剩余的未访问样本集Gamma
            idxGamma = list(set(idxGamma) - set(idxDelta))
    
    # 生成新聚类簇
    k += 1
    idxC_k = list(set(idxOldGamma) - set(idxGamma)) 
    idxC.append(idxC_k) 
    
    # 最后还要剔除同类核心样本
    idxOmega = list(set(idxOmega) - set(idxC_k))

if len(idxGamma)>0:
    # 此算法无法聚类的样本
    k += 1
    idxC.append(idxGamma) 
    idxGamma = []
    
# 根据簇划分idxC，对样本进行分类
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, k)]
for j in range(k):
    X_j = X[idxC[j]]
    plt.plot(X_j[:, 0], X_j[:, 1], 'o', markerfacecolor=colors[j], markeredgecolor='k', markersize=6)
plt.xticks(())
plt.yticks(())
plt.show()
```

## 4）层次聚类

所谓**层次聚类**，就是不停的“自顶往下”地分拆数据集成聚类簇划分， 或不停的“自底往上”地合并单样本集成聚类簇划分。

**AGNES**(AGglomerative NEStiing的简称)是“自底往上”的层次聚类。 核心思想：找出距离最近的两个聚类簇进行合并,如此反复直到给定聚类簇个数。

而**聚类簇间距离**,有如下几种常见形式：

- **最小距离**： $d_{min}(C_i,C_j) = \underset{\boldsymbol{x} \in C_i , \ \boldsymbol{z} \in C_j}{\mathrm{min}} \ d(\boldsymbol{x},\boldsymbol{z})$

- **最大距离**： $d_{max}(C_i,C_j) = \underset{\boldsymbol{x} \in C_i , \ \boldsymbol{z} \in C_j}{\mathrm{max}} \ d(\boldsymbol{x},\boldsymbol{z})$

- **平均距离**： $d_{avg}(C_i,C_j) = \dfrac{1}{|C_i| |C_j|} \sum_{\boldsymbol{x} \in C_i} \sum_{\boldsymbol{z} \in C_j}  d(\boldsymbol{x},\boldsymbol{z})$


算法流程：

![AGNES算法](https://jiage.la/upload/201709213464489451622.png)


下面是Python的实现（样本间采用欧几里得距离， 簇间采用最大距离）：

```python
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt

def dXX(X):
    '''
        计算样本间距离矩阵（采用欧几里得距离）
    '''
    
    # 计算第i个样本到第j个样本的距离D_{i,j}，进而构成m*m距离矩阵D
    # D_{i,j} = (x_i-x_j)^T (x_i-x_j) = x_i^T x_i - 2 x_i^T x_j + x_j^T x_j
    # i = 1,...,m
    # j = 1,...,m
    # D = sqrt(D)
    XX = np.dot(X,X.T)
    D = (XX.diagonal() - 2*XX).T + XX.diagonal()
    return np.sqrt(D)

def dCC(D, idxC):
    '''
        根据样本间距离矩阵D, 计算簇间距离矩阵（采用最大距离）
    '''
    
    nC = len(idxC)
    DC = np.zeros((nC,nC))
    for i in range(nC):
        for j in range(nC):
            if j > i:
                DC[i][j] = D[idxC[i],idxC[j]].max()
                DC[j][i] = DC[i][j]
    return DC
    
    

# 这里选择sklearn提供样本生成器
centers = [[1, 1], [-1, -1], [1, -1]]
X, _ = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=0)
m,n = X.shape

# 预设聚类族个数
k = 3

# 聚类簇划分初始化：每个样本构成一簇  （记录样本索引）
idxC = np.arange(m).reshape(m,1).tolist()

# 计算样本间距离矩阵
D = dXX(X)

# 计算簇间距离矩阵
DC = dCC(D, idxC)
DC[DC==0] = DC.max()+1 # 对后面的计算无影响，但寻找最小距离更方便

# 当前聚类簇个数
q = m

while q > k:
    # 寻找距离最近的连个簇C_i, C_j
    _n = DC.shape[0]
    id = DC.argmin()
    i = id/_n
    j = id%_n
    if i > j:
        jt = j
        j = i
        i = jt
    
    # 合并C_i和C_j到C_i,然后删除C_j
    idxC[i] = list(set(idxC[i]).union(idxC[j]))
    idxC.pop(j)
    
    # 删除簇间距离矩阵的第j行，第j列
    DC = np.delete(DC, j, axis=0)
    DC = np.delete(DC, j, axis=1)
    
    q -= 1
    
    # 重新计算第i行第i列的簇间距离矩阵
    for j in range(q):
        if i != j:
            DC[i][j] = D[idxC[i]][:,idxC[j]].max()
            DC[j][i] = DC[i][j]

    
# 根据簇划分idxC，对样本进行分类
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, k)]
for j in range(k):
    X_j = X[idxC[j]]
    plt.plot(X_j[:, 0], X_j[:, 1], 'o', markerfacecolor=colors[j], markeredgecolor='k', markersize=6)
plt.xticks(())
plt.yticks(())
plt.show()
```
