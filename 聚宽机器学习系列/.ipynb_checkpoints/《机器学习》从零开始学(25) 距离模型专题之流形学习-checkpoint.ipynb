## 8）k近邻学习

【由于后面要用到相关概念，所以将这个最简单的分类器放在这里】

**k近邻学习**（k-Nearest Neighbor, kNN）也是一种常用的监督学习方法。 其工作机制很简单：给定测试样本，基于某种距离度量找出训练集中与其最近的k个训练样本，然后基于这k个“邻居”的信息进行预测。 基于这k个邻居样本，分类任务常用“**投票法**”：k个样本中出现次数最多的类别； 回归任务则常用“**平均法**”：k个样本实值输出的平均值.

这应该是最简单的学习器了。原因是它没有训练过程，属于“懒惰”学习。

下面是这个学习器的Python代码（核心代码只有3行）.

```python
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt

# 这里选择sklearn提供样本生成器
centers = [[1, 1], [-1, -1], [1, -1]]
X, y = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=0)
m,n = X.shape

# 邻居数
k = 3

# 生成网格点对(测试集)
_xx, _yy = np.meshgrid(np.linspace(-3, 3, 50),
                     np.linspace(-3, 3, 50))
TX = np.c_[_xx.ravel(), _yy.ravel()]

# 计算样本集与测试集间距离矩阵
# D_{i,j} = (tx_i-x_j)^T (tx_i-x_j) = tx_i^T tx_i - 2 tx_i^T x_j + x_j^T x_j
# i = 1,...,l
# j = 1,...,m
# l是TX的行数， m是X的行数
D = ((TX**2).sum(axis=1) - 2*np.dot(X,TX.T)).T + (X**2).sum(axis=1)

# 每个网格点离的最近的k个样本(索引)
idxNN = D.argsort()[:,:k+1]  

# 每个网格点最近k个样本出现最多的标签作为预测值（投票法）
ty = np.array(map(lambda v:np.argmax(np.bincount(v)), y[idxNN])) 

# 排成图片方阵
ty = ty.reshape(_xx.shape)

# 绘制网格点所属分类图
plt.imshow(ty, interpolation='nearest',
           extent=(_xx.min(), _xx.max(), _yy.min(), _yy.max()), aspect='auto',
           origin='lower', cmap=plt.cm.Purples_r) #plt.cm.Paired_r) #)

# 按给定标签绘制散点图
plt.scatter(X[:, 0], X[:, 1], s=30, c=y, cmap=plt.cm.Paired,edgecolors='k')
plt.xticks(())
plt.yticks(())
plt.show()
```

## 9）流形学习

**流形学习**这种降维方法，借鉴了拓扑流行的概念。 **流形**，简单地说（非严格说法），就是局部像欧氏空间，但整体不一定像。 我们需要在流形所在高维空间局部地建立到低维空间的映射，然后想办法把局部映射关系推广到全局（至少是样本集的全局）。

### 等度量映射

如果我们假设，局部投影映射后（当然也从局部看）， 任意两个点相距足够近的距离和投影前的距离相等。 我们就称这个投影映射为**等度量映射**（Isometric Mapping, Isomap）。 

进一步，如果我们选择欧氏距离， 计算局部欧氏距离也是投影前的局部距离，但任意两点的距离（最短距离，也就是测地线），是不能简单用欧氏距离直接计算，需要对最短距离路径各节点间欧氏距离求和而来。  

于是我们必须定义“**局部**”，比如： 样本的**k近邻**距离采用欧氏距离计算； 还需要定义“**非局部**”， 比如：样本的**非k近邻**距离设置为无穷大。

那么任意两点A和B的距离，就可能存在一条从A到B的**最短路径**（取决于**局部**定义的大小）， 该路径的各段距离和就是A到B的**最短距离**。 所有**点对**间的最短距离构成等度量映射下的**距离矩阵**。

我们常用图论思路求解这个**最短路径**和**最短距离**， 可以采用著名的**Dijkstra算法**或**Floyd算法**。

下面的代码就是著名的**Dijkstra算法**的Python实现。

```python
from heapq import heappush, heappop
from copy import deepcopy

def relax(W, u, v, D, P):
    '''
        松弛技术
    
    检查是否有有更短路径
    
    参数：
        W[u][v]  对应 u节点到v节点的权重（原始距离，无穷大则表示无关联）
        u 当前节点
        v 下一个节点
        D[u]  初始节点到u节点的最小距离
        P[v]  v节点的先导节点
    '''
    
    # 如果进过节点u到节点v的最小距离d 小于 直接到节点v的最小距离，则修改节点v的最小距离为d，且修改节点v的先导节点为u
    d = D.get(u,np.inf) + W[u][v]
    if d < D.get(v,np.inf):
        D[v], P[v] = d, u
        return True
    return False

def bellman_ford(G, s):
    '''
        检查是否存在负环
    '''
    
    D, P = {s:0}, {}
    for rnd in G:
        changed = False
        for u in G:
            for v in G[u]:
                if relax(G, u, v, D, P):
                    changed = True
        if not changed: break
    else:
        raise ValueError(u"存在负环")
    return D, P

def dijkstra(G, s):
    '''
        Dijkstra最短路径算法
    
    从指定节点出发，试图寻找到其它所有各点的最短距离
    
    参数：
        G 含边权重图。两重字典存储，形如：
            G = {
                0: {1:1.2, 2:0.6},
                1: {0:4.0},
                2: {1:6.0}
            }
        s 出发节点
    返回： D, P
        D: 字典存储  D[j]存储节点s到节点j的最短距离
        P: 字典存储  P[j]表示节点j路先导节点（以节点s为起点）
    '''
    
    # 初始化： 起点到s最短距离，先导节点； 尚未处理的节点； 已处理的节点
    D, P, Q, S = {s:0},{}, [(0,s)], set()
    
    while Q:
        _, u = heappop(Q)
        if u in S: continue
        S.add(u)
        for v in G[u]:
            relax(G, u, v, D, P)
            heappush(Q, (D[v], v))        
    return D, P

def johnson(G):
    '''
        Johnson算法
        
    利用Dijkstra最短路径算法，试图寻找任意两个节点之间的最短路径
    
    参数：
        G 含边权重图。两重字典存储，形如：
            G = {
                0: {1:1.2, 2:0.6},
                1: {0:4.0},
                2: {1:6.0}
            }
    返回： D, P
        D: m*m矩阵  D[i][j]存储节点i到节点j的最短距离
        P: 两重字典存储  P[i][j]表示节点j路先导节点（以节点i为起点）
    '''
    
    # 防破坏
    G = deepcopy(G)
    
    m = len(G)    
    D = np.ones((m,m)) * np.inf
    D[np.diag_indices(m)] = 0
    
    # 添加节点s, 它到所有其它节点0边权值的
    s = object()
    G[s] = {v:0 for v in G}
    # 确保没有负环(如果没有负环，意味着h是空字典)
    h, _ = bellman_ford(G, s)
    del G[s] # 复原
    
    for u in G:
        for v in G[u]:
            G[u][v] += h[u] - h[v]
    
    P = {}
    breakNum = 0
    for u in G:
        _D, P[u] = dijkstra(G, u)
        for v in G:
            D[u][v] = _D.get(v, np.inf) + h[v] -h[u]
            if D[u][v] == np.inf:
                breakNum += 1
    
    if breakNum > 0:
        print (u"有%s对节点间存在断链, 可能是邻域范围太小！"%breakNum)
    return D, P
```

利用**Dijkstra算法**，可以计算出**等度量映射后距离矩阵**，继续看代码：

```python
import numpy as np

import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3

from sklearn.datasets.samples_generator import make_swiss_roll

from sklearn.metrics import euclidean_distances

    
# 用sklearn生成瑞士蛋卷数据集
n_samples = 1500
noise = 0.05
X, _ = make_swiss_roll(n_samples, noise)
X = X[:,:3] # 截取我们所需
#X[:, 1] *= .5 # 使其更瘦
m,n = X.shape

# 近邻参数
k = 10  # 对本例而言，<8 会出现最短距离断链现象

# 数据集中心化
_X = X - X.mean(axis = 0)

# 计算样本集欧氏距离矩阵
D = euclidean_distances(_X)

# k邻接列表
N = D.argsort()[:,:k+1] # 包括了样本到自身的距离，所以要+1
# 根据k邻接列表绘制有向图
G = {i:{j:D[i][j] for j in N[i] if j!=i} for i in range(m)}

# 根据Dijkstra最短路径算法，计算测地线距离矩阵
D,P =johnson(G)

# 距离对称化
D = (D + D.T)/2.0

# 最短路径链实例(第1个样本和第2个样本的最短路径链)
s=0
e=1
print (D[s][e], P[s][e])
path = [e]
while path[0]!=s:
    path.insert(0, P[s][path[0]]) 

# 绘制数据
fig = plt.figure()
ax = p3.Axes3D(fig)
ax.view_init(7, -80)
ax.plot(X[:, 0], X[:, 1], X[:, 2], 'o', markerfacecolor='b', markeredgecolor='k', markersize=6)
ax.plot(X[path, 0], X[path, 1], X[path, 2], 'o', markerfacecolor='r', markeredgecolor='k', markersize=6)
ax.plot(X[N[s], 0], X[N[s], 1], X[N[s], 2], 'o', markerfacecolor='y', markeredgecolor='k', markersize=6)
ax.plot(X[N[e], 0], X[N[e], 1], X[N[e], 2], 'o', markerfacecolor='y', markeredgecolor='k', markersize=6)
ax.plot(X[[s,e], 0], X[[s,e], 1], X[[s,e], 2], 'o', markerfacecolor='g', markeredgecolor='k', markersize=12)
plt.title(u"两个大绿点的最近路径（黄点是绿点的%s近邻）"%k)
plt.show()
```

上面是根据原始样本（选定每个样本邻居样本的前提下）获得每个样本的（关于邻居样本的）误差最小的线性表达。 

**线性局部嵌入**假设降维变换后，能够保持这种变换， 也就是说获得的新数据集也必须是： 每个样本的（关于邻居样本的）误差最小的线性表达。 

于是有前面类似的优化问题：

$$
\boldsymbol{Z}^* = \underset{\boldsymbol{Z}}{\mathrm{argmin}} \ \sum_{i=1}^m \left\| \boldsymbol{z}_i - \sum_{j=1}^k w_{i,j} \boldsymbol{z}_j^{(i)}  \right\|^2
$$

其中: $\boldsymbol{Z}_{m \times d} = (\boldsymbol{z}_1^T;\dots;\boldsymbol{z}_m^T)$。

为了简化表示，我们做如下扩张：

$$
\boldsymbol{W}_{m \times k} \rightarrow [\boldsymbol{W}_{m \times k}, \boldsymbol{0}_{m \times (m-k)}] \overset{\Delta}{=} \boldsymbol{W}_{m \times m} 
$$

于是可简化为：

$$
\boldsymbol{Z}^* = \underset{\boldsymbol{Z}}{\mathrm{argmin}} \ \mathrm{Tr}(\boldsymbol{Z}^T \boldsymbol{M} \boldsymbol{Z}), \quad \boldsymbol{M} = (\boldsymbol{I}-\boldsymbol{W})^T (\boldsymbol{I}-\boldsymbol{W})
$$

我们还希望降维后的“属性”尽可能互不相关，于是有了协方差约束条件：$\boldsymbol{Z}^T \boldsymbol{Z} = \boldsymbol{I}$，于是获得最后的优化问题：

$$
\boldsymbol{Z}^* = \underset{\boldsymbol{Z}}{\mathrm{argmin}} \ \mathrm{Tr}(\boldsymbol{Z}^T \boldsymbol{M} \boldsymbol{Z}), \quad s.t. \boldsymbol{Z}^T \boldsymbol{Z} = \boldsymbol{I}
$$

容易获得这个优化问题最优解所满足的方程：

$$
\boldsymbol{M} \boldsymbol{Z} = \lambda \boldsymbol{Z}
$$

也就是说， 对$\boldsymbol{M}$做特征值分解，其最小的$d$个非零特征值对应的特征向量组成的矩阵就是$\boldsymbol{Z}$。

至此，这个算法完整了。

```python
import numpy as np
from scipy import linalg

import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3

from sklearn.datasets.samples_generator import make_swiss_roll

from sklearn.metrics import euclidean_distances
from sklearn.cluster import KMeans

    
# 用sklearn生成瑞士蛋卷数据集
n_samples = 1500
noise = 0.05
X, _ = make_swiss_roll(n_samples, noise)
X = X[:,:3] # 截取我们所需
#X[:, 1] *= .5 # 使其更瘦
m,n = X.shape

# 数据集中心化
_X = X - X.mean(axis = 0)

# 计算样本集欧氏距离矩阵
D = euclidean_distances(_X)

'''
    本例关注点：开始
'''

# 近邻参数
k = 20  # 对本例而言，<8 会出现最短距离断链现象
# 低维空间维数
d = 2

# k邻接列表
N = D.argsort()[:,:k+1] # 包括了样本到自身的距离，所以要+1

# 计算线性表达权重W_{m \times k}
W = np.zeros((m,k))
for i in range(m):
    # 邻域样本索引
    idx = list(N[i])
    idx.remove(i)
    
    # 邻域样本矩阵
    NX = X[idx]
    
    # 前面数学推导中的Q_i的转置，及其对应的 C_i
    Q = X[i] - NX
    C = np.dot(Q,Q.T)
    C += 1e-6 * C.trace() * np.eye(k)  # 防止出现奇异矩阵
    invC = np.linalg.inv(C)
    
    #第i个样本的线性表达权重
    W[i] = invC.sum(axis=1)/invC.sum()

# 将W扩展成方阵(以0填充)
W = np.concatenate((W,np.zeros((m,m-k))),axis=1)

# 前面数学推导中的M矩阵
MT = np.eye(m)-W
M = np.dot(MT.T,MT)

# 对M进行特征值分解
lam_star, Z_star =linalg.eig(M)

# 对lam从小到大排序(索引)
idxLam = np.argsort(lam_star.real)

print (u"特征值矢量(最小10个): ",list(lam_star[idxLam])[:10])

# 降维，取最小的两个非零特征值 （矩阵M的秩=m-1）
lam = lam_star[idxLam[1:d+1]]
Z = Z_star[:,idxLam[1:d+1]]

if (np.abs(lam.imag)< 0.001).prod():
    lam = lam.real
if (np.abs(Z.imag)< 0.001).prod():
    Z = Z.real

'''
    本例关注点：结束
'''

# 这里我们的目标是降维。至于降维后的聚类，不妨直接调用机器学习算法包（采用k均值聚类）
y_pred = KMeans(n_clusters=9, random_state=170).fit_predict(Z)

# 最后绘制降维后的数据集上的分类
plt.scatter(Z[:, 0], Z[:, 1], c=y_pred)
plt.title(u"在降维空间中的样本数据的分类")
plt.show()

# 绘制原始数据集上的分类
fig = plt.figure()
ax = p3.Axes3D(fig)
ax.view_init(7, -80)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y_pred)
plt.title(u"原始样本数据的聚类")
plt.show()
```
