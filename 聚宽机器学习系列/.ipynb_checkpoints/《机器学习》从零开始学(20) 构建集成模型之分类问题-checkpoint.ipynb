本节用**随机森林**、**梯度提升法**、**以随机森林为基的随机梯度提升法**，解决**二分类问题**（探测水雷）和**多分类问题**（对玻璃进行分类）。

除了展示用Python集成方法工具包解决分类问题，还介绍了**分层采样技术**，用于解决分类数据不均衡的采样问题。

## 3) 用Python集成方法解决二分类问题

### 用随机森林方法探测水雷

本例关注点：

1） 使用了工具包: sklearn.ensemble.RandomForestClassifier

2） 采用了概率预测的预测版本： model.predict_proba(xTest)

3） 采用ROC/AUC来度量性能 （详见代码，并可另外参考第5节和第11节关于ROC/AUC的介绍）

4） 可训练结果包括属性的重要性

5） 体验：AUC越大越好； 随机森林只减少方差，不会减少偏差（不会过拟合）。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import roc_auc_score, roc_curve

target_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data"
df = pd.read_csv(target_url, header=None, prefix='V')

# 分类标签数值化
df['V60'] = df.iloc[:,-1].apply(lambda v:1.0 if v=='M' else 0.0)
# 切分属性和标签
xData = df.values[:,:-1];yData = df.values[:,-1]
m,n = xData.shape
```

```python
# 采样30%的数据作为测试集
xTrain, xTest, yTrain, yTest = train_test_split(xData, yData, test_size=0.30, random_state=531)

auc = []
ntreelist = range(50,2000,50)
for itree in ntreelist:
    depth = None# 允许决策树深度持续增长
    maxfeat = 8# 属性集数目，可调整
    rocksVMinesRFModel = ensemble.RandomForestClassifier(n_estimators=itree,max_depth=depth,max_features=maxfeat,oob_score=False)
    rocksVMinesRFModel.fit(xTrain,yTrain)
    
    # 预测概率
    prediction = rocksVMinesRFModel.predict_proba(xTest)
    
    # 收集测试集的AUC
    auccalc = roc_auc_score(yTest,prediction[:,1:2])
    auc.append(auccalc)

print("AUC")
print(auc[-1])

# 绘制AUC曲线（关于决策树数目）
plt.plot(ntreelist,auc)
plt.xlabel(u'集成模型决策树的数目')
plt.ylabel(u'ROC曲线下面积(AUC)')
#plt.ylim([0.0, 1.1*max(mseOob)])
plt.show()
# 属性的重要性
featureImportance = rocksVMinesRFModel.feature_importances_
featureImportance = featureImportance / featureImportance.max()

# 绘制最重要的头30个属性
idxsorted = np.argsort(featureImportance)[30:60]
idxtemp = np.argsort(featureImportance)[::1]#可能是反转序列
print(idxtemp)
barpos = np.arange(idxsorted.shape[0])+0.5
plt.barh(barpos,featureImportance[idxsorted],align='center')
plt.yticks(barpos, df.columns[idxsorted])
plt.xlabel(u'变量重要性')
plt.show()

# 绘制ROC曲线
fpr, tpr, thresh = roc_curve(yTest, list(prediction[:,1:2]))
ctClass = [i*0.01 for i in range(101)]

plt.plot(fpr, tpr, linewidth=2)
plt.plot(ctClass, ctClass, linestyle=':')
plt.xlabel(u'假正率')
plt.ylabel(u'真正率')
plt.show()

# 选取3个特定位置
idx25 = int(len(thresh) * 0.25)
idx50 = int(len(thresh) * 0.50)
idx75 = int(len(thresh) * 0.75)

# 计算总例，正例， 负例
totalPts = len(yTest)
P = sum(yTest)
N = totalPts - P

print ('')
print (u'不同阈值的混淆矩阵')

#  继续沿用前面的数据


# 采样30%的数据作为测试集
xTrain, xTest, yTrain, yTest = train_test_split(xData, yData, test_size=0.30, random_state=531)

# 构建模型
nEst = 2000
depth = 3
learnRate = 0.007
maxFeatures = 20 # 总共20属性，这里就选择了20个属性，本质和设置为None没区别
rockVMinesGBMModel = ensemble.GradientBoostingClassifier(
                                n_estimators=nEst, # 决策树的数目
                                max_depth=depth,  # 决策树的深度
                                learning_rate=learnRate,  # 学习率
                                max_features=maxFeatures) # 默认值None,对应简单决策树，如果<9则表示以随机森林做基学习器
# 训练
rockVMinesGBMModel.fit(xTrain, yTrain)

# 根据梯度提升发产生的一系列模型，生成对应的一系列预测(转换成概率之前的值)
predictions = rockVMinesGBMModel.staged_decision_function(xTest)

# 计算测试集上的AUC
auc = []
aucBest = 0.0
for p in predictions:
    aucCalc = roc_auc_score(yTest, p)
    auc.append(aucCalc)

    # 获取最佳预测
    if aucCalc > aucBest:
        aucBest = aucCalc
        pBest = p

idxBest = auc.index(max(auc))

print(u"最佳AUC" )
print(auc[idxBest])
print(u"最佳AUC所对应的集成规模")
print(idxBest)

# 绘制AUC曲线
plt.figure()
plt.plot(range(1, nEst + 1), rockVMinesGBMModel.train_score_, label=u'训练集偏差', linestyle=":")
plt.plot(range(1, nEst + 1), auc, label=u'测试集AUC')
plt.legend(loc='upper right')
plt.xlabel(u'集成方法规模')
plt.ylabel(u'偏差 & AUC')
plt.show()

# 绘制属性重要性TOP30
featureImportance = rockVMinesGBMModel.feature_importances_
featureImportance = featureImportance / featureImportance.max()

idxSorted = np.argsort(featureImportance)[30:60]

barPos = np.arange(idxSorted.shape[0]) + .5
plt.barh(barPos, featureImportance[idxSorted], align='center')
plt.yticks(barPos, df.columns[idxSorted])
plt.xlabel(u'变量重要性')
plt.show()

# 绘制ROC曲线
fpr, tpr, thresh = roc_curve(yTest, list(pBest))
ctClass = [i*0.01 for i in range(101)]

plt.plot(fpr, tpr, linewidth=2)
plt.plot(ctClass, ctClass, linestyle=':')
plt.xlabel(u'假正率')
plt.ylabel(u'真正率')
plt.show()

# 选取3个特定位置
idx25 = int(len(thresh) * 0.25)
idx50 = int(len(thresh) * 0.50)
idx75 = int(len(thresh) * 0.75)

# 计算总例，正例， 负例
totalPts = len(yTest)
P = sum(yTest)
N = totalPts - P

print ('')
print (u'不同阈值的混淆矩阵')

#25th
TP = tpr[idx25] * P; FN = P - TP; FP = fpr[idx25] * N; TN = N - FP
print ('')
print ('阈值 =   ', thresh[idx25])
print ('真正率(TP) = ', TP/totalPts, '假正率(FP) = ', FP/totalPts)
print ('假负率（FN） = ', FN/totalPts, '真负率(TN) = ', TN/totalPts)

#50th
TP = tpr[idx50] * P; FN = P - TP; FP = fpr[idx50] * N; TN = N - FP
print ('')
print ('阈值 =   ', thresh[idx25])
print ('真正率(TP) = ', TP/totalPts, '假正率(FP) = ', FP/totalPts)
print ('假负率（FN） = ', FN/totalPts, '真负率(TN) = ', TN/totalPts)

#75th
TP = tpr[idx75] * P; FN = P - TP; FP = fpr[idx75] * N; TN = N - FP
print ('')
print ('阈值 =   ', thresh[idx25])
print ('真正率(TP) = ', TP/totalPts, '假正率(FP) = ', FP/totalPts)
print ('假负率（FN） = ', FN/totalPts, '真负率(TN) = ', TN/totalPts)
```

### 用梯度提升方法探测水雷

本例关注点：

1） 使用了工具包: sklearn.ensemble.GradientBoostingClassifier

2） 产生所属类别的实数估计值（转换成概率之前的值）的可迭代版本： model.model.staged_decision_function(xTest)

4） 可训练结果包括属性的重要性

5） 梯度提升法可能会过拟合（会减少偏差），事实上，梯度提升法就是把偏差作为优化目标。

## 4) 用Python集成方法解决多分类问题

### 用随机森林对玻璃进行分类

本例关注点：

1） 分层采样技术：对每一个标签都采样相同比例测试集。 这是解决分类不均衡问题的常用手段。

2） 将分类标签序号化。

3） 使用误分率进行性能评价（多分类问题，如果用AUC将特别复杂）

4） 用工具包方法生成混淆矩阵：confusion_matrix(yTest, pList)

```python
from sklearn.metrics import accuracy_score, confusion_matrix

target_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data"
df = pd.read_csv(target_url, header=None, prefix="V")
df.columns = ['Id','RI','Na', 'Mg', 'Al', 'Si','K', 'Ca', 'Ba', 'Fe', 'Type']
df = df.set_index("Id")

# 切分属性和标签
xData = df.values[:,:-1]; yLabel = df.values[:,-1] 
m,n = xData.shape

# 标签值序号化
labelList = list(set(yLabel))
labelList.sort()
nlabels = len(labelList)
yData = np.array([labelList.index(label) for label in yLabel])
m,n
```

```python
# 分层采样：对每一个标签都采样相同比例测试集
for iLabel in range(nlabels):
    #按标签分层
    idx = [i for i in range(m) if yData[i]==iLabel]
    
    # 采样30%的数据作为测试集
    _xTrain, _xTest, _yTrain, _yTest = train_test_split(xData[idx], yData[idx], test_size=0.30, random_state=531)
    
    if iLabel==0:
        xTrain, xTest, yTrain, yTest = _xTrain, _xTest, _yTrain, _yTest 
    else:
        xTrain= np.append(xTrain,_xTrain, axis=0) 
        xTest= np.append(xTest,_xTest, axis=0) 
        yTrain= np.append(yTrain,_yTrain, axis=0) 
        yTest= np.append(yTest,_yTest, axis=0) 
missCLassError = []
nTreeList = range(50, 2000, 50)
for iTrees in nTreeList:
    depth = None # 允许决策树深度持续增长
    maxFeat  = 4  # 属性集数目，可试这调整
    glassRFModel = ensemble.RandomForestClassifier(
                                n_estimators=iTrees, # 决策树的数目
                                max_depth=depth,  # 决策树的深度
                                max_features=maxFeat, # 属性集数目(如果设置为None,则等同于Bagging方法)
                                oob_score=False, # 不使用样本外估计
                                random_state=531)# 随机数生成器的种子 (在真实的模型训练阶段可设置为None)
    glassRFModel.fit(xTrain,yTrain)

    # 预测概率
    prediction = glassRFModel.predict(xTest)
    
    # 收集测试集错分率
    correct = accuracy_score(yTest, prediction)
    missCLassError.append(1.0 - correct)

print(u"错分率" )
print(missCLassError[-1])

# 生成混淆矩阵
pList = prediction.tolist()
confusionMat = confusion_matrix(yTest, pList)
print('')
print(u"混淆矩阵")
print(confusionMat)

# 绘制错分率曲线（关于决策树数目）
plt.plot(nTreeList, missCLassError)
plt.xlabel(u'集成模型决策树的数目')
plt.ylabel(u'错分率')
#plt.ylim([0.0, 1.1*max(mseOob)])
plt.show()

# 属性的重要性
featureImportance = glassRFModel.feature_importances_
featureImportance = featureImportance / featureImportance.max()

# 绘制最重要的头30个属性
idxSorted = np.argsort(featureImportance)
barPos = np.arange(idxSorted.shape[0]) + .5
plt.barh(barPos, featureImportance[idxSorted], align='center')
plt.yticks(barPos, df.columns[idxSorted])
plt.xlabel(u'变量重要性')
plt.show()
```

### 用梯度提升法对玻璃进行分类

本例关注点：

1） 本例实际上是**以随机森林为基学习器的随机梯度提升法**（maxFeatures=3，subSamp=0.5 详见代码注释）。

2） 前面提及**梯度提升法**会有过拟合，但**随机梯度提升法**没有过拟合。

```python
# 构建模型
nEst = 500
depth = 3
learnRate = 0.003
maxFeatures = 3 # 总共9属性，这里就选择了3个属性，表面以随机森林做基学习器
subSamp = 0.5 # <1.0 表示是随机梯度提升算法。
glassGBMModel = ensemble.GradientBoostingClassifier(
                                n_estimators=nEst, # 决策树的数目
                                max_depth=depth,  # 决策树的深度
                                learning_rate=learnRate,  # 学习率
                                max_features=maxFeatures, # 默认值None,对应简单决策树，如果<9则表示以随机森林做基学习器
                                subsample=subSamp) # 默认1，表示普通梯度提升算法，如果<1，表示是随机梯度提升算法。
# 训练
glassGBMModel.fit(xTrain, yTrain)

# 根据梯度提升发产生的一系列模型，生成对应的一系列预测(转换成概率之前的值)
predictions = glassGBMModel.staged_decision_function(xTest)

# 计算测试集上的错分率
missClassError = []
missClassBest = 1.0
for p in predictions:
    missClass = 1.0*(p.argmax(axis=1)!=yTest).sum()/len(yTest)
    missClassError.append(missClass)

    # 获取最佳预测
    if missClass < missClassBest:
        missClassBest = missClass
        pBest = p

idxBest = missClassError.index(min(missClassError))

print (u"最佳误分率") 
print  (missClassBest)
print (u"最佳误分率所对应的集成规模")
print (idxBest)

# 绘制误分率曲线
missClassError = [100*mce for mce in missClassError]
plt.figure()
plt.plot(range(1, nEst + 1), glassGBMModel.train_score_, label=u'训练集偏差', linestyle=":")
plt.plot(range(1, nEst + 1), missClassError, label=u'测试集误差')
plt.legend(loc='upper right')
plt.xlabel(u'集成方法规模')
plt.ylabel(u'偏差 & 误分率')
plt.show()

# 绘制属性重要性
featureImportance = glassGBMModel.feature_importances_
featureImportance = featureImportance / featureImportance.max()

idxSorted = np.argsort(featureImportance)
barPos = np.arange(idxSorted.shape[0]) + .5
plt.barh(barPos, featureImportance[idxSorted], align='center')
plt.yticks(barPos, df.columns[idxSorted])
plt.xlabel(u'变量重要性')
plt.show()

# 绘制最佳预测的混淆矩阵
bestPrediction= pBest.argmax(axis=1)
confusionMat = confusion_matrix(yTest, bestPrediction)
print ('')
print (u"混淆矩阵")
print (confusionMat)
```
