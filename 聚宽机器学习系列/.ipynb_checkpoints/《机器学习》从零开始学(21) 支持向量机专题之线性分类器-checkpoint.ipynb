**支持向量机**（SVM, Support Vector Machine）是在高维特征空间使用线性函数假设空间的学习系统。这里涉及两个关键点：1）线性分类器；2）映射到高维线性特征空间后，依旧使用线性分类器。 

本节重点在 线性分类器。  逻辑回归 =》 感知机 =》 线性可分&硬间隔SVM =》 线性不可分&软间隔SVM


## 1) 线性分类器

### 从逻辑回归到线性分类器

首先**逻辑回归**就是用来分类的，参见： <a href="https://www.joinquant.com/post/8123">《机器学习》从零开始学(12) 惩罚线性方法构建模型之含惩罚项逻辑回归</a>

**逻辑回归**从形式上看，就是一个线性模型外套一个**Logistic函数**：

$$P(y \ | \ \boldsymbol{x}) = \dfrac{1}{1 + \exp (-  \boldsymbol{x}^T \boldsymbol{\beta})}, \quad \boldsymbol{x}=(1,x_1,\dots,x_n)^T, \quad \boldsymbol{\beta}=(\beta_0,\beta_1,\dots,\beta_n)^T $$.

我们注意到，条件概率$P(y \ | \ \boldsymbol{x})$在$(0,1)$中关于0.5是对称的，恰好映射到$\boldsymbol{x}^T \boldsymbol{\beta}$在$(-\infty,\infty)$中关于0是对称的。  如果我们认为概率大于0.5代表正例，那么意味着$\boldsymbol{x}^T \boldsymbol{\beta} >0$也代表正例记着1；反之，$\boldsymbol{x}^T \boldsymbol{\beta} < 0$代表负例，对称地记作-1。于是写成简单而对称的形式：

$$y=\mathrm{sign} (\boldsymbol{x}^T \boldsymbol{\beta})  \overset{\triangle}{=}  \mathrm{sign} (f(\boldsymbol{x})) $$

而超平面$\boldsymbol{x}^T \boldsymbol{\beta}=0$被称作**决策面**，也被称作**线性分类器**。一个**线性分类器**或**决策面**完全由$\boldsymbol{\beta}$确定。

### 最小二乘分类器与基本线性分类器

对于给定的数据集$\left\{ (\boldsymbol{x}_i,y_i) \right\}_1^m$，最容易想到的一个决策面就是**最小二乘分类器**：

$$
\boldsymbol{\beta} = \boldsymbol{M}^{-1}\boldsymbol{X}^T\boldsymbol{y} = \boldsymbol{M}^{-1} (n^{\oplus} \boldsymbol{\mu}^{\oplus} - n^{\ominus} \boldsymbol{\mu}^{\ominus}), \quad \boldsymbol{M} = \boldsymbol{X}^T \boldsymbol{X}, \quad \boldsymbol{X}=(\boldsymbol{x}_1^T,\dots, \boldsymbol{x}_m^T)^T
$$

其中，$n^{\oplus}$代表正例数；$\mu_j^{\oplus}$代表所有正例的第j属性值均值；负例也有类似的表示。

特别地，我们假设属性是零中性化，属性等方差且互不相关，那么$\boldsymbol{M}$就是等值对角矩阵；进一步假设同类同分布，即$n^{\oplus}=n^{\ominus}$。 在这些假设下，上式可简化为：

$$
\boldsymbol{\beta} = c (\boldsymbol{\mu}^{\oplus} -\boldsymbol{\mu}^{\ominus})
$$

这种$\boldsymbol{\beta}$所谓对应的特殊的决策面，被称作**基本线性分类器**。

一般而言，即使在线性可分的训练集上，**最小二乘分类器**的准确性也不可能100%。

### 线性可分与感知器

如果给定的数据集$\left\{ (\boldsymbol{x}_i,y_i) \right\}_1^m$是**线性可分**，即存在一个**决策面**可以把这个数据集的正负实例完全分开，这个决策面就是一个线性分类器，并且称之为**感知器**。


**最小二乘分类器**或**基本线性分类器**均无法获得完美的决策面。下面的代码可以说明这一点（出现了错分点）：

```python
import numpy as np
import matplotlib.pyplot as plt

# 随机产生一个线性可分的玩具数据集
m = 100
np.random.seed(200)
xx = np.random.sample(2*m).reshape((m,2))*10
np.random.shuffle(xx)
y = np.zeros(m)
f = 3*xx[:,0]+5*xx[:,1]-45
y = y+(f>4)
y = y-(f<-4)
xx = xx[y!=0]
y = y[y!=0]

m,n=xx.shape

# 最小二乘分类器 (手工计算，扩展X成[1,X]方便计算)
_xx = np.concatenate((np.ones((m,1)), xx), axis=1)
beta = np.linalg.inv(_xx.T.dot(_xx)).dot(_xx.T).dot(y)

# 绘制决策面: beta0 + x_1*beta1 + x_2*beta2 = 0
x1 = np.arange(11)*1.0
x2 = -x1*(beta[1]/beta[2]) -(beta[0]/beta[2])
plt.plot(x1,x2)

colors = [('r' if v>0 else 'b') for v in y]
plt.scatter(xx[:,0],xx[:,1], c=colors)
plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```

根据**感知器**定义本身就要求是完美决策面。  实际上，**Novikoff定理**保证至少可以找到一个完美决策面的**感知机学习算法**： 

--------------

从$\boldsymbol{\beta} = 0$开始迭代，在每一次迭代中，遍历所有样本$i=1,\dots,m$, 如果错分，即$y_i \boldsymbol{x}_i^T \boldsymbol{\beta} \le 0$ ，那么更新$\boldsymbol{\beta} \leftarrow \boldsymbol{\beta} + \eta y_i \boldsymbol{x}_i$ ，直到遍历所有样本无更新。 其中，$0<\eta \le 1$是**学习率**。

--------------

下面的代码演示用**感知机学习算法**对前面的玩具数据集进行完美分类。

```python
# 继续沿用前面玩具数据集的扩展 X 和 扩展 \beta

eta = 0.5 # 学习率

# 扩展\beta初始化
beta = np.zeros(n+1)
while True:
    # 遍历所有样本
    all_right = True
    for i in range(m):
        if y[i]* np.dot(_xx[i],beta) <= 0:
            beta = beta + eta*y[i]*_xx[i]
            all_right = False
    if all_right:
        break

# 绘制决策面: beta0 + x_1*beta1 + x_2*beta2 = 0
x1 = np.arange(11)*1.0
x2 = -x1*(beta[1]/beta[2]) -(beta[0]/beta[2])
plt.plot(x1,x2)

colors = [('r' if v>0 else 'b') for v in y]
plt.scatter(xx[:,0],xx[:,1], c=colors)
plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```

### 对偶形式的感知机学习算法

如果每个样本$i$的错分次数为$\alpha_i$（并将**学习率**$\eta$这个常数吸收进去，或取$\eta=1$），那么上述迭代完成后，$\boldsymbol{\beta}$可写成训练实例的线性组合：

$$
\boldsymbol{\beta} = \sum_{i=1}^m { \alpha_i y_i \boldsymbol{x}_i}
$$

这意味着：从对偶形式的、基于实例角度看线性分类，我们所学习的并非特征权值$\beta_j$，而是实例的权值$\alpha_i$。

很自然，实例$\boldsymbol{x}$的预测值$\hat{y}$就可以改写成：

$$\hat{y}=\mathrm{sign} \left( \sum_{i=1}^m {\alpha_i y_i \boldsymbol{x}_i^T \boldsymbol{x} } \right) $$

这意味着：在训练过程中，所需要的关于训练数据的信息只是实例对的内积，而包含这些内积的矩阵$\boldsymbol{G}=\left\{ \boldsymbol{x}_i^T \boldsymbol{x}_j \right\}_{n \times n}$被称为**Gram矩阵**。

根据上面的表述，于是有**对偶形式的感知机学习算法**：

--------------

从$\boldsymbol{\alpha} = 0$开始迭代，在每一次迭代中，遍历所有样本$i=1,\dots,m$, 如果错分，即$ y_i \sum_{j=1}^m {\alpha_j y_j \boldsymbol{x}_i^T \boldsymbol{x}_j} \le 0$ ，那么更新该样本的错分计数$\alpha_i \leftarrow \alpha_i + 1$ ，直到遍历所有样本无更新。 

--------------

下面的代码演示用**对偶形式的感知机学习算法**对前面的玩具数据集进行完美分类。

```python
# 继续沿用前面玩具数据集的扩展 X 和 扩展 \beta

# \alpha初始化
alpha = np.zeros(m)
while True:
    # 遍历所有样本
    all_right = True
    for i in range(m):        
        if (_xx.T*(alpha*y)).sum(axis=1).dot(_xx[i])*y[i] <= 0:
            alpha[i] += 1
            all_right = False
    if all_right:
        break

# 有了\alpha, 就可以算出\beta
beta = (_xx.T*(alpha*y)).sum(axis=1)

# 绘制决策面: beta0 + x_1*beta1 + x_2*beta2 = 0
x1 = np.arange(11)*1.0
x2 = -x1*(beta[1]/beta[2]) -(beta[0]/beta[2])
plt.plot(x1,x2)

colors = [('r' if v>0 else 'b') for v in y]
plt.scatter(xx[:,0],xx[:,1], c=colors)
plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```

## 2) 支持向量机

### 感知机的缺陷与间隔

前面的分析表明，**感知机**能够找到可分数据集的完美决策面。但是这种完美决策面并非唯一的，特别是在数据量少的时候。 

从直观上（想象一下就足够了）看，用完美决策面切割样本后，两边**最近的**样本都离这个完美决策面尽**可能远**。很自然，这样完美决策面是我们最想要的。

**注意**：从下面开始，$\boldsymbol{x}=(x_1,\dots,x_n)^T$和$\boldsymbol{\beta}=(\beta_1,\dots,\beta_n)^T$都不再是扩展向量。

于是引入完美决策面到最近数据的**间隔**(margin)概念（直观上很自然）。在此之前，我们很容易首先想到并算出，任何一个样本$\boldsymbol{x}_i$到决策面$\boldsymbol{x}^T \boldsymbol{\beta}+\beta_0 = 0$的有向距离$\tilde{d}_i$是：

$$
\tilde{d}_i = \dfrac{\boldsymbol{x}_i^T \boldsymbol{\beta} + \beta_0}{\left\| \boldsymbol{\beta} \right\|^2}
$$

我们将有向距离变成绝对距离(在线性可分的前提下)，将样本$\boldsymbol{x}_i$变成最近样本，就形成了完美决策面到最近数据的**间隔**(margin)：

$$
d = \dfrac{ \underset{i}{\mathrm{min}} \left\{y_i \ (\boldsymbol{x}_i^T \boldsymbol{\beta} + \beta_0)\right\}}{\left\| \boldsymbol{\beta} \right\|^2}
$$

当然，如果线性不可分，那么必然有$d<0$。

下面的代码， 根据前面已经算出的$\beta$所对应的完美决策面，计算间隔。

```python
# 沿用前面已经算出决策面

d = (y*_xx.dot(beta)).min()/(np.dot(beta[1:],beta[1:]))
print ("完美决策面到最近数据的间隔:",d)
```


### 最大化间隔

从前面算出决策面来看，直觉上就觉得有更好的选择：最大化间隔就是我们最想要的、唯一的完美决策面：

$$
\beta_0^*,\boldsymbol{\beta}^* = \underset{\beta_0, \ \boldsymbol{\beta}}{\mathrm{argmax}} \left\{\dfrac{ \underset{i}{\mathrm{min}} \left\{y_i \ (\boldsymbol{x}_i^T \boldsymbol{\beta} + \beta_0)\right\}}{\left\| \boldsymbol{\beta} \right\|^2} \right\}
$$

我们注意到，这个优化问题是一个比值， 分子分母同时乘（除）一个非零系数是不影响结果的，于是我们把分子置1，同时将分子变成约束条件。于是上面的优化问题变成含约束的分母最小化问题：

$$
\beta_0^*,\boldsymbol{\beta}^* = \underset{\beta_0, \ \boldsymbol{\beta}}{\mathrm{argmin}} \left\{\dfrac{1}{2}\left\| \boldsymbol{\beta} \right\|^2 \right\}, \quad \underset{i}{\mathrm{min}} \left\{y_i \ (\boldsymbol{x}_i^T \boldsymbol{\beta} + \beta_0)\right\}=1
$$

即归结成一个含约束二次规划问题：

$$
\beta_0^*,\boldsymbol{\beta}^* = \underset{\beta_0,\ \boldsymbol{\beta}}{\mathrm{argmin}} \left\{\dfrac{1}{2}\left\| \boldsymbol{\beta} \right\|^2 \right\}, \quad y_i \ (\boldsymbol{x}_i^T \boldsymbol{\beta} + \beta_0)\ge 1, \quad i=1,\dots,m
$$

### 用拉格朗日乘子法导出对偶优化问题

构造含乘子的拉格朗日函数$L(\beta_0, \boldsymbol{\beta};\boldsymbol{\alpha})$，进而上述优化问题变成：

$$
\beta_0^*,\boldsymbol{\beta}^* = \underset{\beta_0,\boldsymbol{\beta}}{\mathrm{argmin}} \ \underset{\alpha \ge 0}{\mathrm{max}} \ L(\beta_0, \ \boldsymbol{\beta};\boldsymbol{\alpha})
$$

$$
\begin{array}
& L(\beta_0, \boldsymbol{\beta};\boldsymbol{\alpha}) &=& \dfrac{1}{2}\left\| \boldsymbol{\beta} \right\|^2 - \sum_{i=1}^m{\alpha_i \left( y_i (\boldsymbol{x}_i^T \boldsymbol{\beta} + \beta_0) - 1 \right)} \\
&=& \dfrac{1}{2}\left\| \boldsymbol{\beta} \right\|^2 - \left( \sum_{i=1}^m{\alpha_i y_i \boldsymbol{x}_i^T } \right) \boldsymbol{\beta} - \left( \sum_{i=1}^m{\alpha_i y_i} \right) \beta_0 + \sum_{i=1}^m{\alpha_i}
\end{array}
$$

先求最大值再求最小值的形式，可以确保原优化问题的不等式约束条件成立，并且可以求出最佳$\beta_0^*,\boldsymbol{\beta}^*$。

当然我们也可以先求最小值再求最大值，就可以求出最佳$\boldsymbol{\alpha}$：

$$
\boldsymbol{\alpha}^* = \underset{\alpha \ge 0}{\mathrm{argmax}} \ \underset{\beta_0,\boldsymbol{\beta}}{\mathrm{min}} \ L(\beta_0, \ \boldsymbol{\beta};\boldsymbol{\alpha})
$$


首先求关于$\beta_0^*,\boldsymbol{\beta}^*$的最小值，即通过关于$\beta_0^*,\boldsymbol{\beta}^*$的一阶导为零，容易求出：

$$
\boldsymbol{\beta} = \sum_{i=1}^m{\alpha_i y_i \boldsymbol{x}_i}, \quad \sum_{i=1}^m{\alpha_i y_i} = 0
$$

我们注意到，$\beta_0$的值并没有出现在对偶问题中，利用原始约束容易找到$\beta_0$：

$$
\beta_0 = -\dfrac{1}{2} 
\left(\underset{y_i=-1}{\mathrm{max}} \left\{\boldsymbol{x}_i^T \boldsymbol{\beta}\right\} + \underset{y_i=1}{\mathrm{min}} \left\{\boldsymbol{x}_i^T \boldsymbol{\beta}\right\}\right)
$$

然后再求最大值，即关于$\boldsymbol{\alpha}$的、原优化问题的**对偶优化问题**(含约束)：


$$
\boldsymbol{\alpha}^* = \underset{\boldsymbol{\alpha}}{\mathrm{argmax}} \ \left\{ - \dfrac{1}{2} \sum_{i,j=1}^m {\alpha_i \alpha_j y_i y_j \boldsymbol{x}_i^T \boldsymbol{x}_j}  + \sum_{i=1}^m {\alpha_i} \right\}, \quad \sum_{i=1}^m{\alpha_i y_i}=0, \quad \alpha_i \ge 0, \quad i=1,\dots,m
$$

一旦解出了$\boldsymbol{\alpha}$, $\beta_0,\boldsymbol{\beta}$也就出来了，进而可得到模型：

$$y=\mathrm{sign} \left( \sum_{i=1}^m {\alpha_i y_i \boldsymbol{x}_i^T \boldsymbol{x} } + \beta_0 \right) $$

这个**对偶优化问题**就是**硬间隔支持向量机**算法的全部。 所谓**硬间隔**就是数据集线性可分的**间隔**。

和前面的**感知机**特性类似，寻找最大间隔决策面等价于寻找支持向量（即那些非零拉格朗日乘子所对应的训练样本）。其次，该优化问题完全由训练实例的逐对内积（即Gram矩阵）所确定。

关于这个SVM优化问题有更快的算法-SMO算法（见下一篇笔记）。

下面的代码是直接调用数学算法包进行计算（手工代码强算有点麻烦）,计算结果直观上就已经很满意了：

```python
import scipy.optimize as opt

# 求解硬间隔SVM

init_alpha = np.zeros(m)

bds = [(0,None) for i in range(m)]
cons = {"type":"eq","fun":lambda v:np.dot(v,y)}

def fun(_alpha):
    _beta = (xx.T*(_alpha*y)).sum(axis=1)
    return np.dot(_beta,_beta)/2.0-_alpha.sum()

sol = opt.minimize(fun, init_alpha, method='SLSQP', bounds=bds, constraints=cons)
if sol['status']==0:
    print ("计算成功")

# 获取最优\alpha值
alpha = sol["x"]

# 有了\alpha, 就可以算出\beta
beta = (xx.T*(alpha*y)).sum(axis=1)

# 计算beta0
tmp = np.dot(xx,beta)
beta0 = -(tmp[y>0].min() + tmp[y<0].max()) /2.0

# 绘制决策面: beta0 + x_1*beta1 + x_2*beta2 = 0
x1 = np.arange(11)*1.0
x2 = -x1*(beta[0]/beta[1]) -(beta0/beta[1])
plt.plot(x1,x2)

colors = [('r' if v>0 else 'b') for v in y]
plt.scatter(xx[:,0],xx[:,1], c=colors)
plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```

## 3) 软间隔支持向量机

前面讨论的是线性可分的情况，但实际数据往往是不可分的，可能是噪音导致的。 **硬间隔**的最大问题是：它总是完美地产生一个没有训练误差的一致假设。  当数据不能完全分开时，这个**间隔**是负值。

为了允许在一定程度上违反间隔约束，需要引入$m$个松弛变量$\left\{\xi_i\right\}_1^m$表示**间隔误差**，进而构建优化问题（以一阶范数软间隔为例）：

$$
\beta_0^*,\boldsymbol{\beta}^*,\boldsymbol{\xi}^* = \underset{\beta_0,\ \boldsymbol{\beta},\ \boldsymbol{\xi}}{\mathrm{argmin}} \left\{\dfrac{1}{2}\left\| \boldsymbol{\beta} \right\|^2 + C \sum_{i=1}^m \xi_i \right\}, \quad y_i \ (\boldsymbol{x}_i^T \boldsymbol{\beta} + \beta_0)\ge 1 - \xi_i, \quad \xi_i \ge 0, \quad i=1,\dots,m
$$

其中，$C$是一个自选参数，用于调节间隔误差的惩罚力度。

和前面类似，两组不等式约束，需要引入两组拉格朗日乘子$\left\{\alpha_i\right\}_1^m$和$\left\{r_i\right\}_1^m$来构造拉格朗日$L(\beta_0, \boldsymbol{\beta}, \boldsymbol{\xi};\boldsymbol{\alpha},\boldsymbol{r})$：

$$
L(\beta_0, \boldsymbol{\beta}, \boldsymbol{\xi};\boldsymbol{\alpha},\boldsymbol{r}) = \dfrac{1}{2}\left\| \boldsymbol{\beta} \right\|^2 - \left( \sum_{i=1}^m{\alpha_i y_i \boldsymbol{x}_i^T } \right) \boldsymbol{\beta} - \left( \sum_{i=1}^m{\alpha_i y_i} \right) \beta_0 + \sum_{i=1}^m {\xi_i(C-\alpha_i-r_i)} + \sum_{i=1}^m{\alpha_i},\quad \alpha_i \ge 0, \quad r_i \ge 0 
$$

采用前面类似的方法，先对$\beta_0, \boldsymbol{\beta}, \boldsymbol{\xi}$依次求偏导数等于0,得到3组方程：

$$
\boldsymbol{\beta} = \sum_{i=1}^m {\alpha_i y_i \boldsymbol{\beta}_i}, \quad \sum_{i=1}^m {\alpha_i y_i } = 0
$$

$$
C - \alpha_i - r_i = 0 \Rightarrow  \alpha_i \le C, \quad i=1,\dots,m
$$

代入到拉格朗日中，和前面类似可得到**对偶优化问题**(含约束)：

$$
\boldsymbol{\alpha}^* = \underset{\boldsymbol{\alpha}}{\mathrm{argmax}} \ \left\{ - \dfrac{1}{2} \sum_{i,j=1}^m {\alpha_i \alpha_j y_i y_j \boldsymbol{x}_i^T \boldsymbol{x}_j}  + \sum_{i=1}^m {\alpha_i} \right\}, \quad \sum_{i=1}^m{\alpha_i y_i}=0, \quad 0 \le \alpha_i \le C, \quad i=1,\dots,m
$$

由此可见，除了对$\boldsymbol{\alpha}$的约束不一样，其他都和**硬间隔**类似。

关于这个软间隔SVM优化问题有更快的算法，依然是SMO算法（见下一篇笔记）。

下面的代码是直接调用数学算法包进行计算（手工代码强算有点麻烦）,计算结果直观上就已经很满意了：

```python
#  重新生成线性不可分的玩具数据集
m = 100
np.random.seed(200)  # 保持和前面基本一致
xx = np.random.sample(2*m).reshape((m,2))*10
np.random.shuffle(xx)
y = np.zeros(m)
f = 3*xx[:,0]+5*xx[:,1]-45
y = y + (f>4)
y = y - (f<-4)
_len = len(y[(f>=-4)&(f<=4)])
y[(f>=-4)&(f<=4)] = np.random.choice([-1,1],_len)
m,n = xx.shape

# 求解软间隔SVM  (和求硬间隔SVM完全类似)

C = 1 # 违反硬间隔的惩罚系数
init_alpha = np.zeros(m)
bds = [(0,C) for i in range(m)]

cons = {"type":"eq", "fun": lambda v: np.dot(v,y)}

def fun(_alpha):
    _beta = (xx.T*(_alpha*y)).sum(axis=1)
    return np.dot(_beta,_beta)/2.0 - _alpha.sum()

sol = opt.minimize(fun, init_alpha, method='SLSQP', bounds=bds, constraints=cons)
if sol['status']==0:
    print ("计算成功")

# 获取最优\alpha值
alpha = sol["x"]

# 有了\alpha, 就可以算出\beta
beta = (xx.T*(alpha*y)).sum(axis=1)

# 计算beta0
tmp = np.dot(xx,beta)
beta0 = -(tmp[y>0].min() + tmp[y<0].max()) /2.0

# 绘制决策面: beta0 + x_1*beta1 + x_2*beta2 = 0
x1 = np.arange(11)*1.0
x2 = -x1*(beta[0]/beta[1]) -(beta0/beta[1])
plt.plot(x1,x2)

colors = [('r' if v>0 else 'b') for v in y]
plt.scatter(xx[:,0],xx[:,1], c=colors)
plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```
