**支持向量机**（SVM, Support Vector Machine）是在高维特征空间使用线性函数假设空间的学习系统。这里涉及两个关键点：1）线性分类器；2）映射到高维线性特征空间后，依旧使用线性分类器。 

本节重点在 核方法与SMO。

## 4）使用核函数的非线性模型

前面解决的线性模型，即使线性不可分，也仅仅是因为比如噪音的原因导致的，本质依然是线性的。 但对本质就是非线性模型的问题，前面的方法就不好使了。

### 引入核函数

回顾前面**感知机**、**硬间隔SVM**、**软间隔SVM**各自对应的对偶算法，都有一个共性： 在训练过程中，所需要的关于训练数据的信息只是**实例对的内积**，而包含这些内积的矩阵$G=\left\{\boldsymbol{x}_i^T \boldsymbol{x}_j \right\}_{n \times n}$ 被称为**Gram矩阵**。

我们可以把这个共性进一步抽象为：

$$
\boldsymbol{x}_i^T \boldsymbol{x}_j \overset{抽象}{=} \boldsymbol{\phi}(\boldsymbol{x}_i)^T \boldsymbol{\phi}(\boldsymbol{x}_j) = \kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)
$$

其中，$\boldsymbol{\phi}(\boldsymbol{x})$ 是**输入空间**到**特征空间**的映射，空间维数可以不一样，一般特征空间的维数大于输入空间。对线性问题而言，$\boldsymbol{\phi}$是恒等映射，**特征空间**就是**输入空间**。而$\kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)$就是映射到**特征空间**中两个向量的内积，称作**核函数**。  事实上，无需先计算映射$\boldsymbol{\phi}$，再求内积，只要给出**输入空间**的核函数$\kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)$表达式，就可以在**输入空间**直接计算。

如果原问题映射到**特征空间**中变成线性问题，那么就可以用上节关于线性分类器的解法。 一切都很完美（以一阶范数软间隔SVM为例）：

$$
\boldsymbol{\alpha}^* = \underset{\boldsymbol{\alpha}}{\mathrm{argmax}} \ \left\{ - \dfrac{1}{2} \sum_{i,j=1}^m {\alpha_i \alpha_j y_i y_j \kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)}  + \sum_{i=1}^m {\alpha_i} \right\}, \quad \sum_{i=1}^m{\alpha_i y_i}=0, \quad 0 \le \alpha_i \le C, \quad i=1,\dots,m
$$

$$y=\mathrm{sign} \left( \sum_{i=1}^m {\alpha_i y_i \kappa (\boldsymbol{x}_i,\boldsymbol{x}) } + b \right) $$

其中，$b$对应原来的$\beta_0$，但由于核函数的掩盖，原本有限维的$\boldsymbol{\beta}$已经失去了意义。但$\beta_0$也就是现在的$b$还有意义，那就是**决策面**(可能是曲面)的**阈值**。

### 正定核

在实际应用中，往往依赖领域经验知识直接选择核函数，而核函数的有效性需要通过实验的验证。

但不是任意构造一个$\kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)$都能成为核函数，最基本的要求简单说就是：可以根据$\kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)$，构造出一个希尔伯特空间。

引入希尔伯特空间目的是，尽可能模仿欧几里得空间（事实上，欧几里得空间就一个有限维的希尔伯特空间，而希尔伯特空间可以是无限维的）。除了维数，欧几里得空间具备的重要性质，希尔伯特空间都有。比如：欧几里得空间有内积和长度的概念， 通过$\kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)$构造出的希尔伯特空间，这个$\kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)$本身就是内积，进而可构造出和“长度”类似**范数**的概念。

粗略地说（系统地说有点麻烦，需要讲太多的泛函分析的知识）， 欧几里得空间对长度的要求就是我们对$\kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)$的要求。 我们知道欧几里得空间长度是**对称半正定的**，即对任意$A$和$B$两点而言：

$$
d(A,B) = d(B,A), \quad d(A,A)\ge 0,  \forall A,B
$$

基于希尔伯特空间是欧几里得空间的推广，那么必然要求核函数$\kappa (\boldsymbol{x},\boldsymbol{z})$也是**对称半正定的**，这样的核函数被称作**正定核**。

**注意**： 书中的表述是正确并比较严格的，作为笔记有点啰嗦，  我采用类比、更通俗的方式重新表述之。

### 常用核函数

一旦定义了核函数， 原则上SVM在非线性模型上的应用问题就已经解决了。 

下面给出几种常用的核函数：

【多项式核函数】

$$
\kappa (\boldsymbol{x},\boldsymbol{z}) = \left( \boldsymbol{x}^T \boldsymbol{z} + 1\right)^p
$$

【高斯核函数】

$$
\kappa (\boldsymbol{x},\boldsymbol{z}) = \exp \left(- \dfrac{\| \boldsymbol{x} - \boldsymbol{z} \|^2}{2 \sigma^2}\right)
$$

## 5）序列最小最优化算法（SMO）

SVM中计算量比较大的部分，就是下面这个优化问题：

$$
\boldsymbol{\alpha}^* = \underset{\boldsymbol{\alpha}}{\mathrm{argmin}} \ \left\{ \dfrac{1}{2} \sum_{i,j=1}^m {\alpha_i \alpha_j y_i y_j \kappa (\boldsymbol{x}_i,\boldsymbol{x}_j)}  - \sum_{i=1}^m {\alpha_i} \right\}, \quad \sum_{i=1}^m{\alpha_i y_i}=0, \quad 0 \le \alpha_i \le C, \quad i=1,\dots,m
$$

上一节，我们是采用二次规划的算法包进行求解，但实际训练中是比较慢的。 这里我们介绍一个更高效的算法：**序列最小最优化算法**（sequential minimal optimization, SMO）。

**基本思路**：如果$\boldsymbol{\alpha}$满足此优化问题的KKT条件，那么此优化问题的解就得到了。 否则，选择两个变量，固定其它变量，算出更好解（关键是这是个封闭解）。  这个思路中有如下3个关键点。

### 1. KKT条件

**KKT条件**，就是用拉格朗日乘子法求解含不等式约束优化问题最优解所必须满足的条件。 对上面这个优化问题而言，KKT条件整理后为：

$$
\sum_{i=1}^m{\alpha_i y_i}=0
$$

$$
0 \le \alpha_i \le C, \quad i=1,\dots,m
$$

$$
y_i g(\boldsymbol{x}_i) = \left\{\begin{array}
& \ge 1, \quad \alpha_i = 0 \\
= 1, \quad 0 < \alpha_i < C \\
\le 1, \quad \alpha_i  = C
\end{array} \right . , \quad i=1,\dots,m
$$

其中(b的初值可设为0)：

$$
g(\boldsymbol{x}_i) = \sum_{j=1}^m {\alpha_j y_j \kappa (\boldsymbol{x}_j,\boldsymbol{x}_i)} + b
$$


### 2. 两个变量的选择

一旦选择了2个变量，一个是违反KKT最严重的那个，另一个则由等式约束自动确定。

**第一个变量**$\alpha_i$，就是违反下面这个条件最严重的：

$$
y_i g(\boldsymbol{x}_i) = \left\{\begin{array}
& \ge 1, \quad \alpha_i = 0 \\
= 1, \quad 0 < \alpha_i < C \\
\le 1, \quad \alpha_i  = C
\end{array} \right . , \quad i=1,\dots,m
$$

即，下面$i$所对应的变量$\alpha_i$即为当前所选:

$$
i = \underset{i}{\mathrm{argmax}}  \left| y_i g(\boldsymbol{x}_i)-1 \right|, \quad s.t. \quad y_i g(\boldsymbol{x}_i)不满足KKT条件
$$

**第二个变量**$\alpha_j$，就是在等式约束下，能够使$\alpha_j$有最大变化（参考后面两变量二次规划问题的求解结果）：

$$
j = \underset{j}{\mathrm{argmax}}  \left| E_j -E_i \right|, \quad  E_k = g(\boldsymbol{x}_k) - y_k
$$


### 3. 两变量二次规划问题求解

一旦确定了当前2个变量，不失一般性，不妨指$\boldsymbol{\alpha}_1,\boldsymbol{\alpha}_2$这两个变量， 于是前面的优化问题变成了：

$$
\alpha_1^*,\alpha_2^* = \underset{\boldsymbol{\alpha}}{\mathrm{argmin}} \ \left\{ \dfrac{1}{2} \alpha_1^2 y_1^2 \kappa_{1 \ 1} + \dfrac{1}{2} \alpha_2^2 y_2^2 \kappa_{2 \ 2} + \alpha_1 \alpha_2 y_1 y_2 \kappa_{1 \ 2}  - (\alpha_1+\alpha_2) + \alpha_1 y_1 \sum_{j=3}^m {\alpha_j y_j \kappa_{1 \ j}} + \alpha_2 y_2 \sum_{j=3}^m {\alpha_j y_j \kappa_{2 \ j}} + \dfrac{1}{2} \sum_{i,j=3}^m {\alpha_i \alpha_j y_i y_j \kappa_{i \ j}}  \right\}, \quad  \alpha_1 y_1 + \alpha_2 y_2 = -\sum_{i=3}^m{\alpha_i y_i} = \zeta, \quad 0 \le \alpha_i \le C, \quad i=1,2
$$

这个优化问题，不等式约束使得解在一个正方形区域内，等式约束使得问题变成一个单变量优化问题， 原则上应该有一个封闭解，并且不难求解，仅仅有点繁琐而已。 求解思路就是先求**无不等式约束**的解， 如果解满足**不等式约束**，求解完成，如果不满足，再用**不等式约束**进行裁剪，获得在**不等式约束**边界上的解。

下面直接写出求解结果：

- $\alpha_2$在不等式约束下的上下边界：


$$
H = \left\{\begin{array}
& \min (C, C + \alpha_2^{old} - \alpha_1^{old}), \quad y_1 \ne y_2 \\
\min (C, \alpha_2^{old} + \alpha_1^{old}), \quad y_1 = y_2
\end{array} \right .
$$


$$
L = \left\{\begin{array}
& \max (0, \alpha_2^{old} - \alpha_1^{old}), \quad y_1 \ne y_2 \\
\max (0, \alpha_2^{old} + \alpha_1^{old} - C), \quad y_1 = y_2
\end{array} \right .
$$

- $\alpha_2$无不等式约束的解：

$$
\alpha_2^{new,unc} = \alpha_2^{old} + \dfrac{y_2 (E_1 - E_2)}{\eta}, \quad \eta = \kappa_{1 \ 1} + \kappa_{2 \ 2} - 2 \kappa_{1 \ 2}
$$

- $\alpha_1, \alpha_2$含不等式约束的解：


$$
\alpha_2^{new} = \left\{\begin{array}
& H, \quad \alpha_2^{new,unc} > H \\
\alpha_2^{new,unc}, \quad 0 \le \alpha_2^{new,unc} \le H \\
L, \quad \alpha_2^{new,unc} < L 
\end{array} \right .
$$

$$
\alpha_1^{new} = \alpha_1^{old} + y_1 y_2 (\alpha_2^{old} - \alpha_2^{new}) 
$$

最后，再补上一个$b$的更新值：

$$
b1 = -E_1 - y_1 \kappa_{1,1} (\alpha_1^{new} - \alpha_1^{old}) - y_2 \kappa_{2,1} (\alpha_2^{new} - \alpha_2^{old}) + b
$$

$$
b2 = -E_2 - y_1 \kappa_{1,2} (\alpha_1^{new} - \alpha_1^{old}) - y_2 \kappa_{2,2} (\alpha_2^{new} - \alpha_2^{old}) + b
$$

$$b^{new} = \dfrac{b1+b2}{2}$$

至此，SMO算法已经完整了。
